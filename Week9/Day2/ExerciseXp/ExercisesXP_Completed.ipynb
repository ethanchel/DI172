### Your answer

**Model Choice**

I would select **Logistic Regression** as the baseline model and **Random Forest** as the primary model for loan default prediction. Random Forest is chosen as the primary model because it handles non-linear relationships between features, is robust to outliers (which are common in income and loan amount data), naturally provides feature importance rankings, and generally achieves higher predictive accuracy than linear models without requiring extensive feature engineering.

**Evaluation Plan**

The data would be split into 80% training and 20% testing sets, with stratification applied to ensure that the class ratio of defaulted to non-defaulted loans is preserved in both splits. To obtain a reliable estimate of model performance and avoid overfitting to a single data split, I would apply **Stratified 5-Fold Cross-Validation** on the training set during model development. The primary metric I would report is the **F1-score**, because it balances precision and recall and is more informative than accuracy when the classes are imbalanced. I would also report **ROC-AUC** to measure the model's overall ability to discriminate between classes across all thresholds, and **PR-AUC (Average Precision)** because it is especially informative when the positive class (defaults) is the minority. If the imbalance were more severe, I would additionally apply **SMOTE (Synthetic Minority Over-sampling Technique)** on the training folds only to generate synthetic examples of the minority class, being careful to apply it after the train-test split to prevent data leakage into the test set.

**Unsupervised Learning — K-Means Clustering (Customer Segmentation)**

To assess the effectiveness of a K-Means clustering model, I would use two complementary diagnostic approaches: the **Elbow Method** and the **Silhouette Score**. The Elbow Method involves fitting K-Means for a range of cluster counts (e.g., k = 2 to 10) and plotting the Within-Cluster Sum of Squares (WCSS, also called inertia) against the number of clusters; I would look for the value of k at which the rate of WCSS reduction sharply slows down, forming an "elbow" shape in the plot, and interpret that point as the optimal number of clusters because adding more clusters beyond that point yields diminishing returns in cluster compactness. The **Silhouette Score** would then be computed for the chosen k value: it measures how similar each data point is to its own cluster compared to the nearest neighboring cluster, with values ranging from -1 (misclassified) to +1 (well-clustered), and a score above 0.5 is generally considered indicative of meaningful cluster structure. I would additionally visualize the clusters using a 2D scatter plot generated through PCA dimensionality reduction, which allows for a qualitative inspection of how well-separated the clusters are in the reduced feature space. The key evaluation challenge for unsupervised models is that there is **no ground truth label** to compare against, meaning that all quality metrics are internal (based on the data itself) and the practical meaningfulness of clusters ultimately requires human domain expertise to validate.

**Reinforcement Learning — Q-Learning Agent (Maze Navigation)**

To assess **convergence**, I would monitor the number of steps the agent takes to exit the maze across episodes — a converging agent should show a steady decrease in steps-to-exit that eventually stabilizes near the optimal path length, indicating that the Q-value table has reached a stable equilibrium.
