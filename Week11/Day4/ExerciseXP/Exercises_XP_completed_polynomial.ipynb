## 🌟 Exercise 1: Deep Learning vs Traditional Machine Learning

**As stated in the exercise**  
Objective. Understand the differences between traditional machine learning and deep learning.  
Instructions. Create a table comparing Feature Engineering, Data Processing, Scalability, Pattern Discovery, and Computational Requirements. Identify one real world problem where traditional ML is better, and one where deep learning is better. Write a short paragraph explaining why deep learning has an advantage in unstructured data processing.

**Guidance**  
Use a markdown table for the comparison, keep wording concise and concrete, then write two short examples and one paragraph in full sentences.


| Aspect | Traditional ML | Deep Learning |
|:--|:--|:--|
| Feature Engineering | Requires manual, domain-expert-driven feature extraction before model training | Learns relevant features automatically from raw data through multiple layers |
| Data Processing | Performs well with small to medium structured/tabular datasets | Requires large volumes of data to reach peak performance; struggles with small datasets |
| Scalability | Does not scale well with very large datasets; performance often plateaus | Performance continues to improve as more data and compute are added |
| Pattern Discovery | Limited to relatively shallow, human-defined patterns in the feature space | Discovers hierarchical, abstract patterns automatically (edges → shapes → objects) |
| Computational Requirements | Low to moderate; runs efficiently on standard CPUs without specialized hardware | High; typically requires GPUs or TPUs and significant memory for training |

**Real-world problem examples:**

**Traditional ML — Credit Risk Scoring:** Predicting whether a bank customer will default on a loan is better handled by traditional ML (e.g., Gradient Boosted Trees). The dataset is structured and tabular (income, credit history, debt ratio), the number of samples is moderate, and the model needs to be highly interpretable for regulatory compliance — areas where traditional ML excels.

**Deep Learning — Medical Image Diagnosis:** Detecting tumors in radiology scans (e.g., chest X-rays or MRIs) is better handled by a Convolutional Neural Network. The input is high-dimensional unstructured pixel data, and the spatial hierarchies of features (tissue textures, shapes, abnormalities) cannot be hand-engineered effectively — exactly the setting where deep learning thrives.


**Why deep learning excels at unstructured data:**

Deep learning has a fundamental advantage in unstructured data processing because it performs **representation learning** directly from raw inputs, eliminating the need for manual feature engineering. Through its stacked layers, a deep neural network progressively builds **hierarchical features** — for example, a convolutional network learns low-level edges in early layers, mid-level shapes in intermediate layers, and high-level semantic concepts ("cat face", "tumor boundary") in deeper layers. Traditional ML models cannot discover these latent hierarchies automatically; they depend on human experts to define and compute features from the raw data first, which is time-consuming, error-prone, and often impossible for complex modalities like audio waveforms or natural language. Because deep learning learns both the features and the decision boundary jointly in an end-to-end fashion, it can exploit subtle statistical regularities in images, text, and audio that would be invisible to hand-crafted feature pipelines. This is why deep learning has become the dominant approach for computer vision, speech recognition, and natural language processing.


## 🌟 Exercise 2: Artificial Neural Networks (ANNs)

**As stated in the exercise**  
Objective. Identify and understand the key components of an ANN.  
Instructions. Draw or use a diagram tool to create an ANN with an input layer of 3 neurons, one hidden layer of 4 neurons, and an output layer of 2 neurons. Label neurons, weights, biases, activation function, and layers. Write a short description of how information flows through the network.

**Guidance**  
A clear schematic is preferred. Ensure every requested label appears exactly once and is readable.


```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

fig, ax = plt.subplots(figsize=(12, 7))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.axis('off')
ax.set_facecolor('#f8f9fa')
fig.patch.set_facecolor('#f8f9fa')

# ── layer x positions ──────────────────────────────────────────
x_input  = 2.0
x_hidden = 5.0
x_output = 8.0

# ── neuron y positions ─────────────────────────────────────────
def y_positions(n, total_height=8, offset=1):
    """Evenly space n neurons vertically."""
    step = total_height / (n + 1)
    return [offset + step * (i + 1) for i in range(n)]

y_in  = y_positions(3)
y_hid = y_positions(4)
y_out = y_positions(2)

radius = 0.38

# ── draw weights (connections) ─────────────────────────────────
# input → hidden
for yi in y_in:
    for yh in y_hid:
        ax.plot([x_input + radius, x_hidden - radius], [yi, yh],
                color='#adb5bd', lw=0.9, zorder=1)

# hidden → output
for yh in y_hid:
    for yo in y_out:
        ax.plot([x_hidden + radius, x_output - radius], [yh, yo],
                color='#adb5bd', lw=0.9, zorder=1)

# ── label one weight on each side ──────────────────────────────
mx1 = (x_input + x_hidden) / 2
my1 = (y_in[0] + y_hid[-1]) / 2
ax.text(mx1, my1 + 0.25, 'w₁₂  (weight)', fontsize=8.5,
        color='#495057', ha='center',
        bbox=dict(boxstyle='round,pad=0.2', fc='white', ec='#adb5bd', lw=0.7))

mx2 = (x_hidden + x_output) / 2
my2 = (y_hid[0] + y_out[-1]) / 2
ax.text(mx2, my2 + 0.25, 'w₂₃  (weight)', fontsize=8.5,
        color='#495057', ha='center',
        bbox=dict(boxstyle='round,pad=0.2', fc='white', ec='#adb5bd', lw=0.7))

# ── draw neurons ───────────────────────────────────────────────
neuron_styles = [
    (x_input,  y_in,  '#4361ee', 'x'),   # input
    (x_hidden, y_hid, '#7209b7', 'h'),   # hidden
    (x_output, y_out, '#f72585', 'ŷ'),   # output
]

for xc, ys, color, sym in neuron_styles:
    for i, yc in enumerate(ys):
        circle = plt.Circle((xc, yc), radius,
                             color=color, zorder=3, ec='white', lw=1.5)
        ax.add_patch(circle)
        ax.text(xc, yc, f'{sym}_{i+1}', fontsize=9, color='white',
                ha='center', va='center', fontweight='bold', zorder=4)

# ── bias labels (one per layer receiving input) ────────────────
for xc, ys, label in [(x_hidden, y_hid, 'b₁ (bias)'), (x_output, y_out, 'b₂ (bias)')]:
    ax.annotate(label,
                xy=(xc, ys[0] - radius),
                xytext=(xc, ys[0] - 1.1),
                fontsize=8.5, color='#6c757d', ha='center',
                arrowprops=dict(arrowstyle='->', color='#6c757d', lw=0.9))

# ── activation function label ──────────────────────────────────
ax.text(x_hidden, y_hid[-1] + 0.9,
        'Activation\nfunction: ReLU(z)',
        fontsize=8.5, color='#7209b7', ha='center',
        bbox=dict(boxstyle='round,pad=0.3', fc='#f3e8ff', ec='#7209b7', lw=0.8))

# ── layer title labels ─────────────────────────────────────────
for xc, title, color in [
    (x_input,  'INPUT LAYER\n(3 neurons)',  '#4361ee'),
    (x_hidden, 'HIDDEN LAYER\n(4 neurons)', '#7209b7'),
    (x_output, 'OUTPUT LAYER\n(2 neurons)', '#f72585'),
]:
    ax.text(xc, 9.4, title, fontsize=9.5, ha='center', va='center',
            color='white', fontweight='bold',
            bbox=dict(boxstyle='round,pad=0.4', fc=color, ec='none'))

ax.set_title('Artificial Neural Network — 3 → 4 → 2 Architecture',
             fontsize=13, fontweight='bold', pad=12)
plt.tight_layout()
plt.show()
```

**Forward Propagation — How information flows through the network:**

Each of the 3 input neurons receives one raw feature value (x₁, x₂, x₃) and passes it forward through weighted connections. In the hidden layer, each of the 4 neurons computes a **weighted sum** of all inputs plus its own **bias** — formally z = Σ(wᵢ · xᵢ) + b — and then applies a non-linear **activation function** (ReLU in this diagram) to produce its output: h = ReLU(z). These 4 hidden activations are then passed to the output layer, where each of the 2 output neurons repeats the same weighted-sum-plus-bias computation and applies its own activation (e.g., Softmax for classification or Sigmoid for binary output). The final result ŷ₁ and ŷ₂ represent the network's predictions, which are compared to the true labels to compute a loss and update all weights and biases during backpropagation.


**Learning point**  
An ANN composes affine transformations with nonlinear activations. The affine part mixes features through weights and biases. The nonlinearity enables modeling of complex decision boundaries.


## 🌟 Exercise 3: Creating the Dataset and Visualizing the Data

**As stated in the exercise**  
Objective. Generate a dataset with noise and visualize it.  
Instructions. Import numpy, matplotlib.pyplot, and mean_squared_error from sklearn.metrics. Create 20 points using y = -x² with Gaussian noise N(0, 0.05). Plot the points with a scatter plot. Split into a training set with the first 12 points and a test set with the last 8 points.


```python
# PREFILLED: just execute
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

np.random.seed(0)
x = np.arange(-1, 1, 0.1)  # 20 points from -1 to <1 step 0.1
print("x shape:", x.shape)
```

```python
# To-Do: generate y with noise using np.random.normal()
y = -x**2 + np.random.normal(0, 0.05, len(x))
print("y shape:", y.shape)
```

```python
# PREFILLED: just execute (scatter plot of the dataset)
plt.figure(figsize=(5,4))
plt.scatter(x, y)
plt.title("Noisy dataset sampled from y = -x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.tight_layout()
plt.show()
```

```python
# To-Do: split into train (first 12 points) and test (last 8 points)
x_train, y_train = x[:12], y[:12]
x_test,  y_test  = x[12:], y[12:]
print("train sizes:", x_train.shape, y_train.shape,
      "test sizes:",  x_test.shape,  y_test.shape)
```

**Learning point**  
Visualizing noisy data helps separate signal from noise. Splitting into train and test allows you to detect when a model overfits noise instead of learning the underlying pattern.


## 🌟 Exercise 4: Fitting Polynomial Models of Different Degrees

**As stated in the exercise**  
Objective: Fit polynomial models of varying degrees and observe overfitting.  

Instructions: Define `polynomial_fit(degree)` that returns the fitted polynomial coefficients from training data. Define `plot_polyfit(degree)` that draws the training set, the test set, and the fitted polynomial curve. Visualize fits for degrees 1, 7, and 11. Observe behavior on training and test sets as degree increases.


```python
def polynomial_fit(degree: int):
    """
    Fit a polynomial of the given degree to the training data.
    Returns a callable np.poly1d object.
    """
    # np.polyfit returns coefficients of the best-fit polynomial
    coeffs = np.polyfit(x_train, y_train, degree)
    # np.poly1d wraps those coefficients into a callable function
    return np.poly1d(coeffs)

# Quick sanity check
p2 = polynomial_fit(2)
print("Degree-2 polynomial:", p2)
```

```python
def plot_polyfit(degree: int):
    """
    Plots the training set, test set, and the polynomial curve
    fitted to the training data for the given degree.
    """
    # 1) Fit polynomial to training data
    poly = polynomial_fit(degree)

    # 2) Dense x-grid for a smooth curve
    x_line = np.linspace(x.min(), x.max(), 300)

    # 3) Evaluate the polynomial on the grid
    y_line = poly(x_line)

    # 4) Calculate RMSE on train and test for the title
    rmse_train = np.sqrt(mean_squared_error(y_train, poly(x_train)))
    rmse_test  = np.sqrt(mean_squared_error(y_test,  poly(x_test)))

    # 5) Plot
    plt.figure(figsize=(6, 4))
    plt.scatter(x_train, y_train, color='steelblue',  label='Train set',  zorder=3)
    plt.scatter(x_test,  y_test,  color='tomato',     label='Test set',   zorder=3)
    plt.plot(x_line, y_line,      color='forestgreen', linewidth=2,
             label=f'Degree {degree} fit')
    plt.title(f'Polynomial degree {degree}\n'
              f'Train RMSE={rmse_train:.4f}  |  Test RMSE={rmse_test:.4f}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.ylim(-1.4, 0.4)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
# To-Do: visualize degrees 1, 7, and 11
plot_polyfit(1)
plot_polyfit(7)
plot_polyfit(11)
```

**Observations:**

- **Degree 1 (Underfitting):** The linear model is too simple to capture the parabolic shape of y = −x². Both train and test RMSE are high — this is **high bias**.
- **Degree 7 (Near-optimal):** The polynomial starts fitting the underlying curve well, with low train RMSE. Test RMSE is also low, showing good generalization.
- **Degree 11 (Overfitting):** The model memorizes every noisy training point and wiggles wildly between them. Train RMSE is near zero, but test RMSE explodes — this is **high variance / overfitting**.

**Learning point**  
Higher degree polynomials can perfectly fit training data yet fail to generalize. This is overfitting. You should always check test error, not just training error.

## 🌟 Exercise 5: Cross-Validation to Find the Optimal Degree

**As stated in the exercise**  
Objective: Use cross-validation to determine the best polynomial degree.  

Instructions: Loop degrees from 1 to 11. For each degree, compute the RMSE on the training and test sets. Store results in a list. Plot RMSE vs degree with a logarithmic y axis. Identify the degree minimizing test RMSE and confirm it matches the true model y=−x².


```python
def rmse(y_true, y_pred):
    """Root Mean Squared Error."""
    return np.sqrt(mean_squared_error(y_true, y_pred))

degrees = range(1, 12)          # degrees 1 through 11
rows    = []                    # each row: (degree, rmse_train, rmse_test)

for d in degrees:
    poly       = polynomial_fit(d)           # fit on training data
    rmse_train = rmse(y_train, poly(x_train))
    rmse_test  = rmse(y_test,  poly(x_test))
    rows.append((d, rmse_train, rmse_test))

# Peek at first 3 rows
print(f"{'Degree':>8} {'RMSE Train':>12} {'RMSE Test':>12}")
print("-" * 34)
for r in rows:
    print(f"{r[0]:>8}  {r[1]:>11.6f}  {r[2]:>11.6f}")
```

```python
# Pre-filled: plotting RMSE curves (one figure, two curves)
degs    = [r[0] for r in rows]
rmse_tr = [r[1] for r in rows]
rmse_te = [r[2] for r in rows]

# Find the best degree (minimises test RMSE)
best_idx    = int(np.argmin(rmse_te))
best_degree = degs[best_idx]
best_rmse   = rmse_te[best_idx]

plt.figure(figsize=(7, 5))
plt.plot(degs, rmse_tr, marker='o', label='Train RMSE', color='steelblue', linewidth=2)
plt.plot(degs, rmse_te, marker='s', label='Test RMSE',  color='tomato',    linewidth=2)
plt.axvline(best_degree, color='forestgreen', linestyle='--', linewidth=1.5,
            label=f'Optimal degree = {best_degree}')
plt.scatter([best_degree], [best_rmse], color='forestgreen', s=120, zorder=5)
plt.yscale('log')
plt.xlabel('Polynomial degree', fontsize=12)
plt.ylabel('RMSE (log scale)', fontsize=12)
plt.title('RMSE vs Polynomial Degree', fontsize=13, fontweight='bold')
plt.xticks(degs)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\n✅ Best polynomial degree: {best_degree}  (test RMSE = {best_rmse:.6f})")
```

**Analysis — Optimal degree and the bias–variance trade-off:**

The degree that minimises test RMSE is **2**, which corresponds exactly to the true underlying model y = −x². This makes sense: a degree-2 polynomial has just enough capacity to represent a quadratic curve without absorbing the random Gaussian noise in the training data. Degrees below 2 suffer from **high bias** — the model is too simple to capture the parabolic shape, so both train and test errors remain high. Degrees above 2 suffer from **high variance** — the extra polynomial terms chase the noise in the 12 training points, driving train RMSE almost to zero while test RMSE grows sharply. The sweet spot at degree 2 is where the model's complexity matches the true signal, balancing bias and variance to achieve the best generalisation on unseen data — precisely what the bias–variance trade-off framework predicts.
