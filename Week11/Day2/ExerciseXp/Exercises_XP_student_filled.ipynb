**To-Do:** Write your answers below.

### Answers

**1. What is the key difference between traditional machine learning and deep learning?**  
Traditional machine learning relies on manually engineered features that are fed into algorithms such as decision trees or SVMs. Deep learning, by contrast, automatically learns hierarchical representations directly from raw data through multiple layers of a neural network. For example, a traditional ML pipeline for image classification requires hand-crafted edge detectors and histogram features, whereas a deep CNN learns those features on its own during training.

**2. How do artificial neural networks mimic the human brain?**  
ANNs are loosely inspired by the biological brain: artificial neurons receive weighted inputs, sum them up, and pass the result through an activation function — analogous to how a biological neuron fires when its membrane potential exceeds a threshold. Neurons are organized in layers (input, hidden, output), similar to how the cortex processes information through successive regions. The connections between neurons have learnable weights, much like synaptic strengths that change with experience.

**3. Why does deep learning perform better on large datasets compared to traditional machine learning?**  
Deep learning models have millions of parameters and can capture very complex, non-linear patterns. On small datasets they tend to overfit, but as data volume grows they continue to improve their accuracy because more examples allow them to learn finer distinctions. Traditional ML models plateau earlier because their limited capacity and hand-crafted features cannot exploit additional data beyond a certain point. For example, a logistic regression on ImageNet saturates around 70% accuracy, while a ResNet reaches over 95%.

**4. What are some challenges of deep learning, and how can they be addressed?**  
Key challenges include: (a) **large data requirements** — addressed with data augmentation and transfer learning; (b) **high computational cost** — addressed with GPUs/TPUs and efficient architectures; (c) **overfitting** — addressed with dropout, regularization, and early stopping; (d) **vanishing/exploding gradients** — addressed with batch normalization, residual connections, and careful weight initialization; (e) **interpretability** — addressed with explainability tools such as SHAP or Grad-CAM.

**5. What is feature engineering, and why is it not needed in deep learning?**  
Feature engineering is the manual process of selecting, transforming, and creating input variables to improve model performance — for instance, computing pixel histograms or frequency spectra before feeding data into a classifier. Deep learning does not require this step because its hidden layers act as automatic feature extractors: early layers learn low-level patterns (edges, textures) and deeper layers combine them into high-level concepts (faces, objects). The network discovers relevant features by itself through gradient-based optimization.

**6. What role do hidden layers play in a deep learning model?**  
Hidden layers are intermediate processing stages between the input and output. Each layer applies a learned linear transformation followed by a non-linear activation, enabling the network to represent increasingly abstract concepts. Without hidden layers a network can only model linear relationships. With multiple hidden layers, the model can approximate any continuous function — the depth allows it to build complex representations in a compositional way.

**7. In an artificial neural network, what is the function of an activation function?**  
An activation function introduces non-linearity into the network's computations. Without it, stacking multiple layers would be equivalent to a single linear transformation, severely limiting the model's expressive power. For example, ReLU (f(x) = max(0, x)) is widely used in hidden layers because it is computationally cheap, avoids the vanishing gradient problem for positive inputs, and allows deep networks to train efficiently. Softmax is used in the output layer to convert raw scores into a probability distribution over classes.

## 🌟 Exercise 2: Building a Simple Perceptron Decision System

**As stated in the instructions**  
You will create a perceptron-based decision system to decide whether you should go outside based on two inputs:
- Temperature in °F
- Rainy where Yes is 1 and No is 0

Given parameters: Temperature weight 0.6, Rain weight 0.4, Bias 2. Compute the weighted sum
$$\text{Weighted Sum} = 0.6\cdot \text{Temperature} + 0.4\cdot \text{Rain} + \text{Bias}$$
Apply a step activation:
- If Weighted Sum > 20, output 1
- If Weighted Sum ≤ 20, output 0

Evaluate:  
Case 1: Temperature 70°F, Rain 0  
Case 2: Temperature 50°F, Rain 1

**Guidance**  
Implement the weighted sum function yourself. Use a reusable step activation. Then compute the two cases and print the decision for each.

**Learning point**  
A perceptron implements a linear decision boundary. The bias shifts the threshold. Use step activation for simple binary rules. Prefer differentiable activations when training with gradient descent.

```python
# PREFILLED: just execute
from typing import Tuple

W_TEMP = 0.6
W_RAIN = 0.4
BIAS   = 2.0

def step_activation(s: float, threshold: float = 20.0) -> int:
    """Return 1 if s > threshold else 0."""
    return 1 if s > threshold else 0
```

```python
# To-Do: implement weighted_sum then use it to compute the two cases
def weighted_sum(temperature_f: float, rain01: int) -> float:
    """Compute s = 0.6*Temperature + 0.4*Rain + Bias."""
    return W_TEMP * temperature_f + W_RAIN * rain01 + BIAS

# Compute decisions for the two cases and print them
case1 = (70, 0)  # Temperature=70, Rain=0
case2 = (50, 1)  # Temperature=50, Rain=1

s1 = weighted_sum(*case1)
s2 = weighted_sum(*case2)
y1 = step_activation(s1)
y2 = step_activation(s2)

print(f"Case 1 | Temp=70°F, Rain=No  | Weighted Sum={s1:.1f} | Decision={'Go outside ✅' if y1 else 'Stay inside ❌'}")
print(f"Case 2 | Temp=50°F, Rain=Yes | Weighted Sum={s2:.1f} | Decision={'Go outside ✅' if y2 else 'Stay inside ❌'}")
print({"case1": {"sum": s1, "decision": y1}, "case2": {"sum": s2, "decision": y2}})
```

### Interpretation of Results

**Case 1 — Temperature = 70°F, Rain = 0:**  
The weighted sum is 0.6 × 70 + 0.4 × 0 + 2 = 42 + 0 + 2 = **44.0**, which is well above the threshold of 20. The step activation outputs **1**, meaning the perceptron recommends going outside. The high temperature dominates the decision, and the absence of rain adds no negative contribution.

**Case 2 — Temperature = 50°F, Rain = 1:**  
The weighted sum is 0.6 × 50 + 0.4 × 1 + 2 = 30 + 0.4 + 2 = **32.4**, also above the threshold of 20. The step activation outputs **1** again — the perceptron still recommends going outside. Even though it is raining, the temperature of 50°F is warm enough to push the weighted sum far above 20. The rain weight (0.4) is too small relative to the temperature contribution to change the outcome.

**Conclusion:** The perceptron suggests going outside in **both cases**. To prevent it from recommending going out in rain, one would either increase the rain weight, lower the threshold, or redesign the activation boundary.

## 🌟 Exercise 3: Building a Simple Neural Network with TensorFlow or Keras

**As stated in the instructions**  
Build a classifier for MNIST with:
- One input layer
- One hidden layer with 128 neurons and ReLU activation
- One output layer with 10 neurons and softmax activation

**Learning point**  
Normalization stabilizes optimization. One-hot encoding matches softmax outputs to target vectors. ReLU accelerates training by avoiding saturation, while softmax maps logits to a probability simplex.

```python
# PREFILLED: just execute
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.datasets import mnist

print("TensorFlow:", tf.__version__)

# Load data
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize to [0,1]
x_train = x_train.astype("float32") / 255.0
x_test  = x_test.astype("float32")  / 255.0

# Flatten will be added in the model; keep images as 28x28 for now
print("Train shape:", x_train.shape, "Test shape:", x_test.shape)
```

```python
# To-Do: one-hot encode labels y_train and y_test using to_categorical
y_train_oh = to_categorical(y_train, num_classes=10)
y_test_oh  = to_categorical(y_test,  num_classes=10)

print("Labels one-hot:", y_train_oh.shape, y_test_oh.shape)
```

```python
# To-Do: Build a sequential model with Flatten, Dense(128, ReLU), Dense(10, softmax)
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),          # 28x28 -> 784
    layers.Dense(128, activation='relu'),           # hidden layer
    layers.Dense(10,  activation='softmax'),        # output layer (10 classes)
], name='MNIST_FC')

model.summary()
```

```python
# To-Do: compile the model with adam, categorical_crossentropy, and accuracy
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# To-Do: train then evaluate
history = model.fit(
    x_train, y_train_oh,
    epochs=10,
    batch_size=128,
    validation_split=0.1,
    verbose=1
)

test_loss, test_acc = model.evaluate(x_test, y_test_oh, verbose=0)
print({"test_loss": float(test_loss), "test_acc": float(test_acc)})
```

## 🌟 Exercise 4: Forward Propagation Calculation

**As stated in the instructions**  
Predict house prices given:
- Inputs: x₁ = 2000 square feet, x₂ = 3 bedrooms
- Weights: w₁ = 0.5, w₂ = 0.7
- Bias: b = 50,000
- Activation: ReLU

Compute z before activation, apply ReLU, and interpret the prediction.

**Learning point**  
ReLU passes positive linear combinations unchanged and zeros out negatives. Bias sets the baseline prediction independent of inputs.

```python
# To-Do: compute forward propagation for the given numbers
x1, x2 = 2000.0, 3.0
w1, w2 = 0.5, 0.7
b = 50000.0

# Compute z = x1*w1 + x2*w2 + b
z = x1 * w1 + x2 * w2 + b

# Apply ReLU: max(0, z)
y = max(0.0, z)

print(f"Pre-activation z  = {z:,.2f}")
print(f"ReLU output  y    = {y:,.2f}")
print({"z": z, "prediction": y})
```

### Interpretation

The pre-activation value is z = 2000 × 0.5 + 3 × 0.7 + 50000 = 1000 + 2.1 + 50000 = **51,002.10**.

Since z is positive, ReLU leaves it unchanged, so the **predicted house price is $51,002.10**.

**Role of input scaling:** The square footage (x₁ = 2000) contributes 1000 to the sum, while the number of bedrooms (x₂ = 3) only contributes 2.1. Because the raw input values are on very different scales, the weight assigned to square footage has a much larger absolute effect. In practice, inputs should be normalized so that no single feature dominates purely due to its magnitude.

**Role of bias:** The bias (b = 50,000) acts as a fixed baseline price regardless of the features — it represents the minimum value a house would have even with zero square footage and zero bedrooms. Without the bias, the model could only predict prices proportional to the inputs, which would be unrealistic.

**Role of ReLU:** ReLU ensures the predicted price is never negative. If a combination of weights and inputs produced a negative sum, ReLU would output 0, preventing nonsensical negative price predictions.

## 🌟 Exercise 5 (optional): Implementing Forward and Backward Propagation in Python

**As stated in the enoncé**  
You will code a simple neural network that performs forward propagation and backpropagation for a regression problem predicting exam scores from study hours.

**Guidance**  
Run the prefilled code. Then modify the learning rate or initial weights and observe changes in the loss and updates.
Provide a short written explanation.

**Learning point**  
Gradient descent updates parameters in the direction of steepest loss reduction.
The learning rate controls step size. Too small slows learning. Too large risks divergence.

```python
# PREFILLED: just execute
import numpy as np

# Initialize input data (features)
x = np.array([4, 80])  # 4 hours studied, previous test score: 80

# Initialize weights and bias
w = np.array([0.6, 0.3])  # Initial weights
b = 10  # Initial bias

# Forward Propagation
def forward_propagation(x, w, b):
    z = np.dot(x, w) + b  # Weighted sum
    return z  # Linear activation (No ReLU here, it's a regression task)

# Compute prediction
y_pred = forward_propagation(x, w, b)
y_true = 85  # Actual exam score

# Compute Loss (Mean Squared Error)
loss = 0.5 * (y_true - y_pred) ** 2

# Compute Gradients
grad_w = -(y_true - y_pred) * x  # Partial derivatives with respect to weights
grad_b = -(y_true - y_pred)      # Partial derivative with respect to bias

# Update Weights and Bias
learning_rate = 0.01
w_new = w - learning_rate * grad_w
b_new = b - learning_rate * grad_b

# Print Results
print("Initial Prediction:", y_pred)
print("Loss:", loss)
print("Updated Weights:", w_new)
print("Updated Bias:", b_new)
```

```python
# Experiment: compare different learning rates
import numpy as np

x      = np.array([4, 80])
y_true = 85

for lr in [0.001, 0.01, 0.1]:
    w = np.array([0.6, 0.3])
    b = 10.0
    losses = []
    for _ in range(50):
        y_pred  = np.dot(x, w) + b
        loss    = 0.5 * (y_true - y_pred) ** 2
        losses.append(loss)
        grad_w  = -(y_true - y_pred) * x
        grad_b  = -(y_true - y_pred)
        w      -= lr * grad_w
        b      -= lr * grad_b
    print(f"LR={lr:.3f} | Loss after 50 steps: {losses[-1]:.4f} | Final prediction: {np.dot(x, w)+b:.2f}")
```

### Explanation

**Why gradient descent reduces the error:**  
The loss function (MSE) measures how far the prediction is from the true value. Its gradient with respect to each weight indicates the direction in which that weight increases the loss. By subtracting a fraction (the learning rate) of the gradient from the current weight, we move the parameters in the direction that decreases the loss. Repeating this process iteratively drives the prediction closer to the true label.

In the example above, the initial prediction is 0.6×4 + 0.3×80 + 10 = 2.4 + 24 + 10 = **36.4**, whereas the true score is **85**. The large error produces large gradients, so the first update significantly shifts the weights toward values that produce a higher prediction.

**Effect of modifying the learning rate:**  
- **LR = 0.001 (too small):** The weights change very slowly. After 50 steps the loss is still large — the model has barely learned.
- **LR = 0.01 (balanced):** The model converges steadily; loss decreases smoothly over iterations.
- **LR = 0.1 (too large):** The steps overshoot the minimum. The loss may oscillate or even diverge instead of converging.

Choosing the right learning rate is crucial: too small wastes computation, too large prevents convergence.

**Learning point**  
Softmax outputs are probabilities over classes. Inspect the confidence values. Visual errors often indicate confusions between similar digits.

```python
# To-Do: visualize a few predictions with matplotlib
import numpy as np
import matplotlib.pyplot as plt

# Ensure we have the model and x_test, y_test from Exercise 3
# Select 18 random samples from x_test
N = 18
idx    = np.random.choice(len(x_test), N, replace=False)
x_vis  = x_test[idx]
y_true_vis = y_test[idx]   # integer labels (0-9)

# Get predictions (probabilities) then take argmax for the predicted class
y_prob = model.predict(x_vis, verbose=0)
y_pred = np.argmax(y_prob, axis=1)

# Plot
cols = 6
rows = int(np.ceil(N / cols))
fig, axes = plt.subplots(rows, cols, figsize=(14, 2.8 * rows))
fig.suptitle('MNIST Predictions — Green = Correct  |  Red = Wrong', fontsize=14, fontweight='bold')

for i, ax in enumerate(axes.flat):
    if i >= N:
        ax.axis('off')
        continue
    ax.imshow(x_vis[i], cmap='gray')
    correct = y_pred[i] == y_true_vis[i]
    color   = 'green' if correct else 'red'
    conf    = y_prob[i, y_pred[i]] * 100
    ax.set_title(
        f"pred={y_pred[i]}  true={y_true_vis[i]}\n{conf:.1f}%",
        fontsize=9, color=color, fontweight='bold'
    )
    ax.axis('off')

plt.tight_layout()
plt.show()

# Summary
correct_count = np.sum(y_pred == y_true_vis)
print(f"Correct: {correct_count}/{N}  ({correct_count/N*100:.1f}%)")
```
