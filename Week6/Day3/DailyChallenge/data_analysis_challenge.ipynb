# Data Science Salary Analysis
This notebook explores the factors influencing salaries in the data science industry.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('ds_salaries.csv')
df.head()
```

## Preprocessing

```python
df_cleaned = df.drop_duplicates()
print(f'Duplicates removed: {df.shape[0] - df_cleaned.shape[0]}')
```

## Analysis by Experience Level

```python
analysis = df_cleaned.groupby('experience_level')['salary_in_usd'].agg(['mean', 'median']).sort_values(by='mean', ascending=False)
analysis
```

## Visualization

```python
plt.figure(figsize=(10, 6))
analysis['mean'].plot(kind='bar', color='teal')
plt.title('Average Salary (USD) by Experience Level')
plt.ylabel('Salary (USD)')
plt.xlabel('Experience Level')
plt.show()
```
