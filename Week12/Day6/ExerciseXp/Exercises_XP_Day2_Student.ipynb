## 🌟 Exercise 1: Traditional vs. Modern NLP: Comparative Analysis

### 1) Tableau comparatif

| Aspect | Traditional NLP | Modern NLP |
|---|---|---|
| **Feature Engineering** | Manuelle : TF-IDF, n-grammes, POS tags, règles linguistiques codées à la main. Nécessite une expertise domaine importante. | Automatique : les représentations sont apprises directement depuis les données brutes via des réseaux de neurones profonds. |
| **Word Representations** | Statiques : chaque mot a un vecteur fixe (Word2Vec, GloVe, one-hot). "bank" (banque) et "bank" (rive) ont le même vecteur. | Contextuelles : le vecteur d'un mot dépend de son contexte (BERT, GPT). "bank" a deux représentations différentes selon la phrase. |
| **Model Architectures** | Superficielles : Naïve Bayes, SVM, régression logistique, HMM, CRF. Peu de couches, capacité limitée de modélisation. | Profondes : Transformers multi-couches avec attention (BERT, GPT-4, T5). Des milliards de paramètres, capable de capter des dépendances longues. |
| **Training Methodology** | Spécifique à la tâche : un modèle distinct entraîné pour chaque tâche (classification, NER, traduction…) depuis zéro. | Pré-entraînement + fine-tuning : un seul grand modèle pré-entraîné sur d'énormes corpus, puis adapté à n'importe quelle tâche avec peu de données. |
| **Key Examples** | Naïve Bayes, SVM, MaxEnt, HMM, CRF, Word2Vec, GloVe | BERT, GPT-3/4, T5, RoBERTa, LLaMA, PaLM, Claude |
| **Advantages** | Interprétable, léger, rapide à entraîner, peu de données, règles explicites contrôlables. | Haute précision sur toutes les tâches NLP, généralisation remarquable, un seul modèle pour de nombreuses tâches, gère l'ambiguïté contextuelle. |
| **Disadvantages** | Faible sur les tâches complexes, fragile hors-domaine, ingénierie des features coûteuse, ne capture pas le contexte. | Coûteux en calcul et en énergie, opaque (boîte noire), risques de biais, nécessite beaucoup de données pour le pré-entraînement. |

---

### 2) Impact sur la scalabilité et l'efficacité

Le passage au NLP moderne a radicalement transformé la scalabilité des applications linguistiques. Premièrement, **le pré-entraînement sur de grandes données générales** permet à un seul modèle de couvrir des dizaines de tâches sans devoir en créer un nouveau pour chacune, réduisant le temps de mise en production de semaines à quelques heures de fine-tuning. Deuxièmement, **l'apprentissage de représentations automatiques** élimine le goulot d'étranglement du feature engineering manuel, qui était non-scalable pour de nouveaux domaines. Troisièmement, **l'architecture Transformer** est massivement parallélisable sur GPU/TPU — contrairement aux RNN séquentiels — permettant de traiter des batches énormes simultanément. Quatrièmement, **le transfer learning** permet d'obtenir d'excellents résultats avec très peu de données étiquetées dans le domaine cible (quelques centaines d'exemples suffisent pour fine-tuner BERT), ce qui rend ces systèmes accessibles même sans grands jeux de données labellisés. Enfin, **les modèles distillés** (DistilBERT, TinyBERT) reprennent l'efficacité de l'approche moderne tout en réduisant le coût d'inférence, rendant le déploiement sur mobile ou en temps réel viable.



## 🌟 Exercise 2: LLM Architecture and Application Scenarios

---

### BERT — Encoder-only, bidirectionnel

**Architecture :**  
BERT (Bidirectional Encoder Representations from Transformers) utilise uniquement la partie **encoder** du Transformer. Il lit la séquence entière dans **les deux directions simultanément** grâce à l'attention bidirectionnelle. Il est pré-entraîné avec deux objectifs : le **Masked Language Modeling (MLM)** — prédire des tokens masqués aléatoirement dans la phrase — et le **Next Sentence Prediction (NSP)** — prédire si deux phrases se suivent. Cette bidirectionnalité permet à BERT de construire des représentations contextuelles riches qui tiennent compte à la fois du contexte gauche et droit de chaque token.

**Application réelle : Détection de fraude dans des e-mails (classification de texte)**  
Un système de détection de phishing qui analyse chaque e-mail entrant et le classe en "légitime", "phishing", ou "spam".

**Pourquoi BERT est adapté :**  
La classification nécessite de *comprendre* le texte dans sa globalité, pas de le *générer*. BERT excelle à encoder le sens d'une séquence complète en un vecteur de classification (token `[CLS]`), car sa bidirectionnalité lui permet de lier des indices contextuels éparpillés dans tout l'e-mail (ex. : "votre compte" + "cliquez ici" + "urgent" → phishing). Il n'a pas besoin de générer du texte, seulement de comprendre.

---

### GPT — Decoder-only, autorégressif

**Architecture :**  
GPT (Generative Pre-trained Transformer) utilise uniquement la partie **decoder** avec une attention **causale** (masquée) : chaque token ne peut voir que les tokens précédents. Il est pré-entraîné avec l'objectif de **Causal Language Modeling (CLM)** — prédire le prochain token étant donné tous les précédents. Cette architecture est optimisée pour la **génération de texte** séquentielle.

**Application réelle : Assistant de code (GitHub Copilot / GPT-4 Codex)**  
Un système qui suggère la suite du code en temps réel à mesure que le développeur tape, en complétant des fonctions, générant des tests, ou expliquant des erreurs.

**Pourquoi GPT est adapté :**  
La génération de code est intrinsèquement séquentielle : on écrit token après token en tenant compte de ce qui précède. L'attention causale de GPT reproduit exactement ce processus. Le modèle n'a pas besoin d'un "contexte futur" — il doit prédire la prochaine ligne logique à partir du contexte précédent, ce que le CLM optimise directement.

---

### T5 — Encoder-Decoder, text-to-text

**Architecture :**  
T5 (Text-To-Text Transfer Transformer) utilise l'architecture **encoder-decoder complète**. L'encodeur traite le texte source et produit des représentations contextuelles ; le décodeur génère le texte cible token par token en se conditionnant sur ces représentations via de la **cross-attention**. La particularité de T5 est son paradigme **"text-to-text"** : toutes les tâches NLP sont formulées comme une transformation de texte en texte (ex. : "translate English to Spanish: Hello" → "Hola").

**Application réelle : Résumé automatique d'articles scientifiques**  
Un système qui prend un article de recherche de 10 pages et génère un résumé structuré de 3-5 phrases capturant les contributions principales.

**Pourquoi T5 est adapté :**  
Le résumé est une tâche de **transformation** : on lit une longue séquence source (encodeur) et on génère une séquence cible différente et plus courte (décodeur). L'encodeur capture la sémantique globale du document entier avec une attention bidirectionnelle ; le décodeur génère le résumé en sélectionnant les informations pertinentes via la cross-attention. GPT ne peut pas "lire" et "réécrire" efficacement ; BERT ne peut pas générer. T5 est le seul des trois conçu pour cette dualité lecture/génération.



## 🌟 Exercise 3: Benefits and Ethics of Pre-training

---

### Cinq bénéfices clés des modèles pré-entraînés

1. **Improved Generalization (Meilleure généralisation)**  
   Un modèle pré-entraîné sur des milliards de phrases a été exposé à une immense variété de structures linguistiques, de domaines et de styles. Cette diversité lui permet de généraliser à des données qu'il n'a jamais vues lors du fine-tuning. Contrairement à un modèle entraîné depuis zéro sur un seul domaine, il ne surfit pas facilement à des patterns spécifiques du jeu de données cible.

2. **Reduced Need for Labeled Data (Moins de données étiquetées)**  
   Le pré-entraînement est entièrement **auto-supervisé** : le modèle apprend à prédire des tokens masqués ou le prochain mot sans aucune annotation humaine. Seul le fine-tuning nécessite des données étiquetées, et en quantité bien plus faible (quelques centaines à quelques milliers d'exemples suffisent souvent là où une approche from-scratch en exigerait des millions).

3. **Faster Fine-tuning (Fine-tuning plus rapide)**  
   Le modèle possède déjà des représentations linguistiques pertinentes dès le départ. Le fine-tuning n'a besoin que d'ajuster les dernières couches et d'orienter ces représentations vers la tâche cible — quelques heures sur GPU au lieu de semaines. Les embeddings sont déjà "calibrés" sémantiquement, donc la convergence est très rapide.

4. **Transfer Learning (Transfert de connaissances)**  
   Les connaissances acquises sur une tâche ou un domaine se transfèrent à d'autres. Un modèle pré-entraîné sur du texte général peut être fine-tuné pour analyser des comptes-rendus médicaux, des contrats juridiques, ou du code source — sans repartir de zéro. Cette réutilisation des poids est au cœur de l'efficacité du paradigme actuel.

5. **Robustness (Robustesse)**  
   Les modèles pré-entraînés sont plus résistants aux variations de formulation, aux fautes d'orthographe, aux tournures inhabituelles, et au bruit dans les données. Ayant vu des millions de variantes d'une même idée, ils comprennent l'intention derrière un message même imparfaitement formulé.

---

### Préoccupations éthiques du pré-entraînement à grande échelle

- **Biais (Bias) :** Les corpus d'entraînement (web, livres, réseaux sociaux) reflètent les inégalités sociales existantes. Les modèles peuvent reproduire des stéréotypes de genre, de race ou de religion. Ex. : un modèle peut associer "infirmière" à des pronoms féminins et "médecin" à des pronoms masculins.

- **Désinformation (Misinformation) :** Un LLM apprend à générer du texte fluide et convaincant sans mécanisme interne de vérification des faits. Il peut produire des affirmations fausses formulées avec assurance ("hallucinations").

- **Utilisation malveillante (Misuse) :** Les capacités génératives peuvent être exploitées pour créer des deepfakes textuels, du spam personnalisé, de la propagande automatisée, ou des e-mails de phishing indétectables.

- **Vie privée (Privacy) :** Si des données personnelles figurent dans le corpus d'entraînement, le modèle peut mémoriser et restituer ces informations (noms, adresses, numéros de téléphone).

- **Empreinte carbone :** Entraîner GPT-3 a émis autant de CO₂ qu'environ 300 allers-retours New York–San Francisco en avion.

---

### Stratégies d'atténuation

| Problème | Stratégie |
|---|---|
| Biais | Audit des données d'entraînement, débiaisage des embeddings, tests de biais automatisés avant déploiement |
| Désinformation | Ajout d'un module de vérification des faits, fine-tuning avec RLHF (Reinforcement Learning from Human Feedback) |
| Utilisation malveillante | Safety filters, red-teaming, limitation d'accès à l'API, détection de contenu généré (watermarking) |
| Vie privée | Differential Privacy lors de l'entraînement, suppression des données PII avant pré-entraînement |
| Empreinte carbone | Distillation de modèles, entraînement sur des clusters à énergie renouvelable, réutilisation de modèles existants |



## 🌟 Exercise 4: Transformer Architecture Deep Dive

---

### Self-Attention & Multi-Head Attention

**Comment fonctionne le Self-Attention :**  
Pour chaque token d'entrée, on crée trois vecteurs à partir de son embedding : une **Query (Q)**, une **Key (K)** et une **Value (V)**, obtenus par des projections linéaires apprises. Le score d'attention d'un token `i` vers un token `j` est calculé comme le produit scalaire `Qᵢ · Kⱼ`, divisé par `√dₖ` pour stabiliser les gradients. On applique ensuite un **softmax** sur tous les scores pour obtenir une distribution de probabilité (les "poids d'attention"). La sortie finale pour le token `i` est la **somme pondérée des vecteurs V** de tous les tokens, pondérée par ces poids. Ainsi, chaque token "décide" de combien il doit s'intéresser à chaque autre token pour construire sa représentation contextuelle.

**Multi-Head Attention vs. Single-Head :**  
Une seule tête d'attention peut capter un type de relation à la fois. Le multi-head attention effectue `h` attentions en parallèle dans des sous-espaces différents (en projetant Q, K, V dans des espaces de dimension `d_model/h`), puis concatène les résultats. Chaque tête peut se spécialiser dans un type de relation différent : une tête peut capturer les **relations syntaxiques** (sujet-verbe), une autre les **coréférences** (pronoms → antécédents), une autre les **relations sémantiques** (synonymes). Cela permet au modèle de modéliser simultanément plusieurs aspects d'une phrase avec une complexité computationnelle similaire à une seule grande tête.

**Exemple concret — Multi-Head sur une phrase originale :**  
Phrase : *"The trophy didn't fit in the bag because it was too big."*

- **Tête 1 (coréférence)** : le token "it" attend fortement vers "trophy" (le sujet référencé par "it"), l'aidant à résoudre l'ambiguïté.
- **Tête 2 (modification adjectivale)** : "big" attend vers "trophy" pour comprendre qu'il modifie le trophée, pas le sac.
- **Tête 3 (dépendance causale)** : "because" attend vers "fit" et "big" pour lier cause et conséquence.
- **Tête 4 (structure de phrase)** : "didn't" attend vers "fit" pour marquer la négation du verbe.

---

### Objectifs de Pré-entraînement : MLM vs CLM

| | **MLM (Masked LM)** | **CLM (Causal LM)** |
|---|---|---|
| **Principe** | Masquer ~15 % des tokens, prédire les tokens manquants | Prédire le prochain token à partir de tous les précédents |
| **Contexte utilisé** | Bidirectionnel (gauche + droite) | Unidirectionnel (gauche seulement) |
| **Modèles types** | BERT, RoBERTa, ALBERT | GPT, LLaMA, PaLM |
| **Force** | Compréhension profonde du langage | Génération fluide et cohérente |

**Quand préférer MLM :** Pour des tâches de **compréhension** : classification, NER, Q&A extractive, NLI. Ces tâches nécessitent une représentation globale du texte. Ex. : détecter si un avis est positif ou négatif requiert de lire la phrase entière.

**Quand préférer CLM :** Pour des tâches de **génération** : complétion de texte, chatbot, résumé abstractif, traduction. Le modèle doit produire un token à la fois dans un ordre logique. Ex. : un assistant qui répond à une question génère la réponse token par token.

**Pourquoi BERT utilisait NSP et pourquoi les modèles modernes l'abandonnent :**  
Le Next Sentence Prediction (NSP) de BERT original entraînait le modèle à prédire si deux phrases se suivent dans le document. L'idée était d'apprendre des relations inter-phrases utiles pour des tâches comme la question-réponse et l'inférence textuelle (NLI). Cependant, des travaux ultérieurs (notamment RoBERTa, 2019) ont montré que **NSP nuit plus qu'il n'aide** : la tâche est trop facile (les paires négatives viennent de documents distincts et sont facilement distinguables par le domaine), et elle contraint la longueur des séquences d'entraînement à deux courtes phrases au lieu d'un long document. Les modèles modernes préfèrent entraîner sur de **longues séquences continues** uniquement avec MLM, ce qui donne de meilleures représentations.

---

### Sélection de modèle Transformer

**1. Analyse de sentiments sur des avis clients**  
→ **Encoder-only (ex. : BERT, RoBERTa)**  
Justification : la classification de sentiment est une tâche de *compréhension*. On doit lire l'avis en entier et produire une étiquette. Un encodeur bidirectionnel construit une représentation globale de la phrase. Le token `[CLS]` agrège ce contexte global pour la couche de classification. Il n'y a aucun besoin de générer du texte.

**2. Chatbot générant des réponses créatives**  
→ **Decoder-only (ex. : GPT-3/4, LLaMA)**  
Justification : la génération conversationnelle est intrinsèquement **autorégressive** — le modèle produit un token après l'autre en conditionnant chaque nouveau token sur tout ce qui précède (contexte + réponse partielle). L'attention causale du decoder-only est précisément optimisée pour ce schéma. La flexibilité de génération permet d'adapter le style, l'humour, et la créativité.

**3. Traduction de documents techniques (EN → ES)**  
→ **Encoder-Decoder (ex. : T5, mBART, MarianMT)**  
Justification : la traduction est une tâche de **séquence à séquence** — on lit une séquence source (encodeur bidirectionnel, compréhension complète) et on génère une séquence cible dans une autre langue (décodeur autorégressif conditionné via cross-attention). Un encoder-decoder peut aligner des tokens source et cible de longueurs différentes, gérer les inversions de structure grammaticale entre langues, et maintenir la cohérence globale du document traduit.

---

### Encodage Positionnel

**Pourquoi il est essentiel :**  
Le mécanisme d'attention est **invariant à la permutation** : si on mélange l'ordre des tokens, les scores d'attention ne changent pas (car ils ne dépendent que des vecteurs Q/K/V, pas de la position). Pourtant, l'ordre des mots est fondamental en langage. L'encodage positionnel injecte une information de **position absolue ou relative** dans les embeddings pour que le modèle puisse distinguer "Le chat mange la souris" de "La souris mange le chat".

**Exemple de défaillance sans encodage positionnel :**  
Phrase 1 : *"The dog bit the man."*  
Phrase 2 : *"The man bit the dog."*  
Sans encodage positionnel, les deux phrases produisent exactement les mêmes embeddings (les mêmes tokens, juste dans un ordre différent). L'attention verrait les mêmes poids pour "dog" et "man" dans les deux cas, et le modèle serait incapable de distinguer qui mord qui — produisant des prédictions identiques pour des phrases de sens opposé.



## 🌟 Exercise 5: BERT Variations — Choose Your Detective

### Scénarios

**Scénario 1 — Analyse de sentiment en temps réel sur mobile (ressources limitées)**  
→ **DistilBERT**  
DistilBERT est une version distillée de BERT : 40 % plus petit, 60 % plus rapide, tout en conservant 97 % des performances. Sur mobile, la RAM, le CPU et l'autonomie sont critiques. DistilBERT offre le meilleur compromis taille/performance pour un déploiement on-device sans serveur.

**Scénario 2 — Analyse de documents juridiques (haute précision requise)**  
→ **RoBERTa**  
RoBERTa est entraîné bien plus longtemps, sur plus de données, avec de plus grands batches et sans NSP. Ces améliorations lui donnent des représentations plus riches et une précision supérieure à BERT sur presque tous les benchmarks NLP. Pour des documents juridiques où chaque nuance compte, RoBERTa est le bon choix (et encore mieux : Legal-RoBERTa fine-tuné sur des corpus juridiques).

**Scénario 3 — Support client mondial en plusieurs langues**  
→ **XLM-RoBERTa**  
XLM-RoBERTa est pré-entraîné sur 100 langues avec 2,5 TB de texte multilingue. Il supporte le **zero-shot cross-lingual transfer** : fine-tuné sur des données en anglais, il peut classifier du texte en japonais, arabe, swahili, etc. C'est le seul modèle de cette liste conçu pour le NLP multilingue à grande échelle.

**Scénario 4 — Pré-entraînement efficace avec détection de remplacement de tokens**  
→ **ELECTRA**  
ELECTRA remplace le MLM par une tâche de **discriminateur** : un générateur léger remplace certains tokens, et ELECTRA doit détecter quels tokens ont été remplacés (tâche binaire par token). Cette approche est 4x plus efficace en termes de FLOPS que BERT-MLM car chaque token contribue au signal d'apprentissage (pas seulement les 15 % masqués). ELECTRA est précisément la réponse à ce scénario.

**Scénario 5 — NLP efficace dans des environnements contraints**  
→ **ALBERT**  
ALBERT utilise le **parameter sharing** entre couches (les poids du Transformer sont partagés entre tous les blocs encoder) et la **factored embedding parametrization** (séparation de la taille d'embedding et de la taille cachée). Résultat : 18x moins de paramètres que BERT-large avec des performances compétitives — idéal pour des environnements à mémoire limitée.

## 🌟 Exercise 6: Softmax Temperature — The Randomness Regulator

---

### 1) Scénarios de température

La température `T` divise les logits avant le softmax : `p_i = softmax(logits / T)`.  
- **T < 1** → distribution plus pointue (le token le plus probable est encore plus favorisé)  
- **T > 1** → distribution plus plate (les tokens moins probables ont plus de chance d'être sélectionnés)  
- **T = 1** → comportement par défaut du modèle

**T = 0.2 — Distribution très concentrée (déterministe)**  
Le modèle choisit quasi-systématiquement le token ayant le score logit le plus élevé. Les sorties sont très prévisibles, répétitives, et "sûres". Presque aucune variabilité entre deux générations avec le même prompt. Le texte est cohérent et factuel mais peut sembler mécanique, monotone, ou trop prudent. Utile quand on veut de la fiabilité mais on perd toute créativité.

**T = 1.5 — Distribution très plate (aléatoire)**  
Les tokens peu probables ont une chance bien plus grande d'être sélectionnés. Le texte devient plus surprenant, créatif, et varié — mais aussi incohérent, grammaticalement incorrect, ou hors-sujet. Le modèle prend des "risques" linguistiques : il peut introduire des métaphores inattendues, des tournures originales, mais aussi des divagations ou des non-sens. Chaque génération avec le même prompt produira un texte très différent.

**T = 1.0 — Distribution nominale (équilibre)**  
C'est le comportement standard du modèle tel qu'il a été entraîné. La distribution de probabilité reflète exactement ce que le modèle a appris. C'est un bon point de départ : ni trop déterministe ni trop chaotique. La variabilité est présente mais raisonnable.

---

### 2) Application Design

**Histoires de coucher pour enfants — Stratégie de température :**  
Pour des histoires pour enfants, il faut trouver l'équilibre entre **créativité** (pour maintenir l'attention de l'enfant, introduire des personnages surprenants, des rebondissements amusants) et **cohérence** (la narration doit rester compréhensible pour un jeune public).

Je commencerais avec une **température de 0.8–0.9**. Cette plage permet au modèle d'être imaginatif — introduire des animaux parlants, des aventures dans des forêts enchantées — tout en maintenant une structure narrative logique et un vocabulaire adapté. Je n'irais pas au-dessus de 1.0 pour éviter des incohérences narratives (personnages qui disparaissent, retournements incompréhensibles). Je combinrais cette température avec **Top-P (nucleus sampling) à 0.92** pour exclure les tokens vraiment improbables tout en gardant la diversité lexicale.

**Résumés de rapports financiers — Stratégie de température :**  
Pour des rapports financiers, l'exactitude factuelle est non négociable. Une mauvaise synthèse d'un chiffre d'affaires ou d'un ratio financier peut entraîner de mauvaises décisions d'investissement.

J'utiliserais une **température très basse : 0.1–0.2**, combinée avec une stratégie de décodage **greedy ou beam search** (plutôt que du sampling). Cela garantit que le modèle choisit toujours les formulations les plus probables et les plus précises. La créativité n'est pas un objectif ici — la priorité absolue est la fidélité aux données sources.

---

### 3) Température & Biais

**Comment la température affecte l'expression des biais :**

À **basse température**, le modèle amplifie ses choix les plus "confiants" — ceux statistiquement dominants dans les données d'entraînement. Or, si les données d'entraînement contiennent des stéréotypes (ex. : associer certains métiers à un genre ou une ethnie), **une température basse va systématiquement et de façon répétable reproduire ces stéréotypes**, car ce sont les tokens les plus probables. Le biais est concentré, prévisible, et donc potentiellement plus dangereux dans un produit de production.

À **haute température**, le modèle explore davantage de tokens, ce qui peut diluer les associations biaisées les plus fortes — mais peut aussi introduire des biais imprévus ou des incohérences aléatoires.

**Exemple pratique :**  
Prompt : *"The nurse walked into the room and said"* → le modèle complète avec des pronoms.  
- Avec T=0.2 : le modèle génère presque toujours **"she"**, reproduisant systématiquement l'association nurse=femme (stéréotype de genre dans les données d'entraînement).  
- Avec T=1.0 : le modèle génère parfois "he", "they", ou d'autres constructions, rendant le biais moins apparent mais non éliminé.  
- La vraie solution n'est pas d'ajuster la température (c'est un palliatif), mais de **débiaisager les données d'entraînement** ou de faire du **counterfactual data augmentation**.


### (Optional) Quick Generation Demo Across Temperatures
Note: Requires `transformers` and model download; skip if offline.


```python

# ── Démo optionnelle : génération GPT-2 à différentes températures ─────────────
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch

model_name = "gpt2"
prompt = "Artificial intelligence will"

pipe = pipeline(
    "text-generation",
    model=AutoModelForCausalLM.from_pretrained(model_name),
    tokenizer=AutoTokenizer.from_pretrained(model_name),
    device=0 if torch.cuda.is_available() else -1,
)

for T in [0.2, 1.0, 1.5]:
    out = pipe(
        prompt,
        max_new_tokens=40,
        temperature=T,
        do_sample=True,
        pad_token_id=pipe.tokenizer.eos_token_id,  # évite l'avertissement de padding
    )
    print(f"--- Temperature: {T}")
    print(out[0]["generated_text"])
    print()

# Observations attendues :
# T=0.2 → texte répétitif et très "sûr" (ex: "transform industries", "improve healthcare")
# T=1.0 → texte naturel avec une variabilité raisonnable
# T=1.5 → texte plus créatif mais potentiellement incohérent ou grammaticalement hasardeux
```
