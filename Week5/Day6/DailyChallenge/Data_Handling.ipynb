# Daily Challenge: Data Handling and Analysis in Python

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

df = pd.read_csv("datascience_salaries.csv")
df.head()
```

```python
import os
print(os.getcwd())
print(os.listdir())
```

## Min-Max Normalization of Salary

```python

scaler = MinMaxScaler()
df['salary_normalized'] = scaler.fit_transform(df[['salary']])
df[['salary', 'salary_normalized']].head()
```

## Dimensionality Reduction with PCA

```python

numeric_df = df.select_dtypes(include='number').dropna()

pca = PCA(n_components=2)
pca_result = pca.fit_transform(numeric_df)

pca_df = pd.DataFrame(pca_result, columns=['PC1', 'PC2'])
pca_df.head()
```

## Salary Aggregation by Experience Level

```python

salary_stats = df.groupby('experience_level')['salary'].agg(['mean', 'median'])
salary_stats
```
