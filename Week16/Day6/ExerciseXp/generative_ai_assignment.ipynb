# 🌟 Exercise 1: Comparative Analysis of Generative AI and Traditional AI

**Objective:** For each scenario, determine whether Generative AI or Traditional AI is more suitable, with detailed justification.

## 1.1 — Automated Medical Diagnosis

**Scenario:** Detecting lung cancer from X-ray images.

### ✅ Recommended Approach: **Traditional AI (CNN-based classifier)** + Generative AI as support

| Criterion | Traditional AI | Generative AI |
|---|---|---|
| Accuracy | High — CNNs excel at image classification | Lower — generation adds noise risk |
| Interpretability | Moderate (with Grad-CAM) | Low |
| Safety | Validated, deterministic | Hallucination risk |
| Data augmentation | Limited | Strong (can synthesize rare cases) |

### Justification
A **traditional CNN classifier** (e.g., ResNet, EfficientNet) is the primary tool for this task. These models have demonstrated clinical-grade performance on radiology benchmarks (AUC > 0.95). They produce deterministic, reproducible decisions — a hard requirement for regulatory approval (FDA, CE mark).

However, **Generative AI (GANs / Diffusion models)** plays a valuable *supporting* role:
- Synthesizing rare pathology samples to balance training sets.
- Data augmentation without exposing real patient data (privacy preservation).

**Risk of using Generative AI as the primary classifier:** hallucinations in generated intermediate representations could lead to false negatives — a life-threatening outcome in oncology. Traditional AI should be the decision-maker; generative AI is a data pipeline tool.

## 1.2 — Legal Document Generation

**Scenario:** Drafting legally binding contracts with minimal human intervention.

### ✅ Recommended Approach: **Hybrid — Rule-based backbone + Generative AI for drafting**

| Criterion | Rule-Based (Traditional) | Generative AI (LLM) |
|---|---|---|
| Legal precision | Very high | Moderate — hallucination risk |
| Flexibility | Low | High |
| Speed | Fast for templated docs | Fast for novel drafting |
| Auditability | Full | Opaque |
| Cost | Low | Moderate (API calls) |

### Justification
**Rule-based systems** ensure that mandatory clauses (jurisdiction, governing law, liability caps) are always present and correctly formatted. They are deterministic and fully auditable — critical for legal defensibility.

**Generative AI (LLMs)** is excellent at:
- Drafting flexible language for non-standard clauses.
- Summarizing complex documents.
- Adapting tone and structure to context.

### Key Risks of Generative AI in Legal Contexts
1. **Hallucination:** LLMs may fabricate legal precedents, clause numbers, or statute references that don't exist.
2. **Confidentiality breach:** Sensitive contract details may be logged by third-party API providers.
3. **Liability ambiguity:** If an AI-drafted contract causes harm, legal accountability is unclear.

**Recommendation:** Use LLMs for first-draft generation with mandatory human legal review and a rule-based validation layer before signing.

## 1.3 — AI-Generated Scientific Research

**Scenario:** Automating literature reviews and suggesting new research hypotheses.

### ✅ Recommended Approach: **Generative AI — with strong caveats**

### Justification
Generative AI is well-suited for **literature summarization**: it can process thousands of abstracts, extract themes, identify research gaps, and produce structured summaries — tasks that would take human researchers months.

For **hypothesis generation**, generative AI can propose novel connections between concepts, which serves as a creative brainstorming layer. However:

- Generated hypotheses must be validated by domain experts before being taken seriously.
- LLMs may reproduce outdated or retracted findings if trained on stale data.
- There is no guarantee of logical or empirical validity in generated hypotheses.

### Why it works here (vs other scenarios)
Scientific exploration is inherently a creative, exploratory task. Unlike medical diagnosis or legal contracts, an incorrect hypothesis does not cause immediate harm — it gets filtered by the scientific method. The cost of false positives is low; the gain from discovering unexpected connections is high.

**Best implementation:** RAG (Retrieval-Augmented Generation) over a curated corpus of peer-reviewed papers, combined with citation verification.

## 1.4 — Financial Market Predictions

**Scenario:** Predicting stock market trends from historical trading data.

### ✅ Recommended Approach: **Traditional AI (time-series models)** as primary

| Criterion | Traditional AI (LSTM, XGBoost, ARIMA) | Generative AI |
|---|---|---|
| Predictive accuracy | High on structured data | Lower — not designed for forecasting |
| Explainability | Moderate | Very low |
| Regulatory compliance | Easier to audit | Difficult |
| Data synthesis | N/A | Useful for stress-testing |

### Justification
Financial markets are governed by structured, numerical, time-dependent data. Traditional models — **LSTM, Transformer-based time-series (e.g., Temporal Fusion Transformer), XGBoost** — are purpose-built for this and achieve strong performance with clear feature attribution.

**Where Generative AI adds value:**
- **Synthetic scenario generation:** GANs can simulate market crash conditions to stress-test portfolios.
- **Sentiment analysis:** LLMs can parse earnings call transcripts and news to generate signals as input features for traditional models.

**Risk of relying on Generative AI for direct predictions:** LLMs have no causal understanding of market mechanics — they pattern-match language, not financial dynamics. A model that hallucinates a bullish trend could trigger significant financial losses.

## 1.5 — Autonomous Vehicle Decision-Making

**Scenario:** Real-time driving decisions in unpredictable environments.

### ✅ Recommended Approach: **Traditional AI (deterministic + reinforcement learning)** — Generative AI for training only

### Justification
Real-time safety-critical decisions require **deterministic, low-latency, explainable** models. Traditional approaches — perception models (YOLO, PointNet), rule-based planners, reinforcement learning agents — are validated frameworks that offer:
- Predictable behavior in edge cases.
- Hard safety constraints (e.g., never exceed speed limits).
- Regulatory traceability.

Generative AI is **inappropriate as a primary decision-maker** because:
1. It is probabilistic — small prompt changes can produce very different outputs.
2. Latency: LLM inference is too slow for millisecond-scale reactions.
3. No formal safety guarantees exist for generative outputs.

**Where Generative AI helps:**
- Generating synthetic edge-case training data (rare weather, unusual pedestrian behavior).
- Simulating diverse driving environments via world models (e.g., DALL-E + NeRF pipelines).

# 🌟 Exercise 2: Ethical and Security Risks of Generative AI

## 2.1 — Deepfake Political Manipulation

**Scenario:** AI-generated deepfake videos spread misinformation about political candidates.

### Three Key Risks
1. **Democratic integrity:** Voters make decisions based on fabricated evidence — candidates may be shown saying things they never said, altering election outcomes.
2. **Irreversible reputational damage:** Even when debunked, deepfakes create lasting cognitive impressions ("liar's dividend").
3. **Erosion of media trust:** Mass exposure to deepfakes leads to epistemic nihilism — audiences stop trusting any video evidence, including authentic footage.

### Potential Consequences If Unregulated
- Foreign state interference in democratic elections.
- Normalization of disinformation as a political tool.
- Legal systems unable to use video as evidence.

### Two Mitigation Solutions
1. **Technical — Cryptographic content provenance:** Implement the C2PA (Coalition for Content Provenance and Authenticity) standard, embedding tamper-evident digital signatures in all authentic media at the point of capture. Platforms verify signatures before amplifying content.
2. **Regulatory — Pre-election deepfake disclosure laws:** Mandate that any AI-generated political content be clearly labeled. Impose criminal penalties for undisclosed election deepfakes (similar to existing FEC disclosure rules for political ads).

## 2.2 — Synthetic Identity Fraud

**Scenario:** Cybercriminals use AI to create synthetic identities that bypass biometric verification.

### Three Key Risks
1. **Financial system compromise:** Synthetic identities can open bank accounts, take loans, and build credit scores before committing large-scale fraud — losses estimated at $20B+ annually in the US alone.
2. **Biometric system arms race:** As liveness detection improves, so do generative models producing photorealistic faces and voice clones — a perpetual adversarial cycle.
3. **Innocent victim collateral damage:** Legitimate users may be falsely flagged when their real identity is too similar to AI-generated synthetic profiles.

### Potential Consequences If Unregulated
- Collapse of trust in remote KYC (Know Your Customer) processes.
- Systematic fraud against government benefit systems and financial institutions.

### Two Mitigation Solutions
1. **Technical — Multi-modal liveness detection:** Combine behavioral biometrics (typing patterns, mouse movement), depth-sensing cameras, and challenge-response tests (random head movement) that are computationally expensive to fake in real-time.
2. **Systemic — Federated identity verification networks:** Link identity verification across institutions so that synthetic identities attempting to replicate across multiple organizations are flagged by cross-institutional pattern matching.

## 2.3 — Generative AI in Cyber Warfare

**Scenario:** A nation-state generates false intelligence reports to mislead foreign governments.

### Three Key Risks
1. **Military escalation:** A government deceived by fabricated intelligence may take military action based on non-existent threats.
2. **Diplomatic breakdown:** Fabricated diplomatic communications can fracture alliances and destroy years of negotiated trust.
3. **Intelligence community paralysis:** When analysts can no longer distinguish real from synthetic intelligence, decision-making collapses under epistemic uncertainty.

### Potential Consequences If Unregulated
- AI-enabled false-flag operations triggering armed conflict.
- International treaties (e.g., arms control) undermined by fabricated verification data.

### Two Mitigation Solutions
1. **Technical — AI-generated content detection at the intelligence layer:** Deploy classifier ensembles trained specifically on AI-generated text and imagery within intelligence processing pipelines, with mandatory flagging of low-confidence authenticity scores.
2. **Diplomatic — International AI warfare conventions:** Negotiate treaties (analogous to the Geneva Convention for biological/chemical weapons) that prohibit the use of generative AI for creating false intelligence in peacetime, with UN verification mechanisms.

## 2.4 — AI-Generated Malware

**Scenario:** Hacking groups use AI to generate novel malware that evades traditional detection.

### Three Key Risks
1. **Signature evasion at scale:** Traditional antivirus relies on known signatures; AI can generate millions of polymorphic malware variants, overwhelming signature databases.
2. **Democratization of sophisticated attacks:** Script kiddies with no coding skills can now generate zero-day exploits by prompting LLMs.
3. **Accelerated attack cadence:** AI reduces the time to develop, test, and deploy cyberattacks from weeks to hours.

### Potential Consequences If Unregulated
- Critical infrastructure (power grids, hospitals) subjected to unstoppable ransomware campaigns.
- Nation-state cyberattacks that are indistinguishable from criminal activity (attribution collapse).

### Two Mitigation Solutions
1. **Technical — Behavioral AI-based endpoint detection:** Shift from signature-based to behavior-based detection using ML models that identify anomalous process behavior regardless of code signature, making polymorphism irrelevant.
2. **Regulatory — LLM provider-level restrictions:** Implement mandatory safety fine-tuning and output filters in frontier models that refuse to generate functional malware or exploitation code, with third-party auditing of compliance.

## 2.5 — Copyright and Intellectual Property Theft

**Scenario:** A generative AI trained on copyrighted works generates content that closely resembles them.

### Three Key Risks

### Potential Consequences If Unregulated
- Collapse of creative industries unable to compete with zero-marginal-cost AI content.
- Precedent that training data does not require consent, undermining all intellectual property law.

### Two Mitigation Solutions
1. **Legal — Mandatory training data licensing frameworks:** Require AI developers to obtain licenses for training data, with revenue sharing mechanisms for rights holders (similar to music streaming royalty models).
2. **Technical — Style fingerprinting and similarity thresholds:** Implement post-generation filters that detect and block outputs that exceed a defined similarity threshold to copyrighted works, using perceptual hashing and style embedding comparisons.

## 3.2 — Bias and Fairness in AI Training Data

**Problem:** A model trained on biased news sources generates politically skewed content.

### Dataset Curation Strategies

1. **Source diversification:** Build training corpora from a balanced matrix of sources across the political spectrum (left, center, right), international outlets, and non-English sources translated to target language — ensuring no single editorial voice dominates.

2. **Bias auditing with automated tools:** Use tools like **Perspective API**, **AllenNLP bias detectors**, or custom classifiers trained on labeled political bias datasets (e.g., AllSides Media Bias Ratings) to score and filter documents.

5. **Adversarial debiasing during fine-tuning:** Apply **Adversarial Training for Fairness** — train an auxiliary classifier to detect political lean in model outputs, and penalize the generative model when the classifier successfully identifies a lean.

## 3.3 — Fine-Tuning for Domain-Specific Tasks

## 3.5 — Controlling AI Creativity and Coherence

### Temperature Scaling
Temperature `T` is applied to the logit distribution before softmax:
$$P(x_i) = \frac{\exp(z_i / T)}{\sum_j \exp(z_j / T)}$$

- **T < 1:** Distribution sharpens → model picks high-probability tokens → more coherent, repetitive, conservative output.
- **T > 1:** Distribution flattens → model samples less likely tokens → more creative, potentially incoherent output.
- **T = 0:** Greedy decoding — always picks the most probable token.

**Practical use:** Set T ≈ 0.7 for factual tasks, T ≈ 1.2 for creative writing.

### Reinforcement Learning from Human Feedback (RLHF)
1. Collect human preference rankings over model outputs.
2. Train a **reward model** to predict human preferences.
3. Fine-tune the generative model with **PPO (Proximal Policy Optimization)** to maximize reward model score while adding a KL-divergence penalty to prevent drift from the base distribution.

This constrains creativity to the space humans find coherent and useful, while still allowing novel outputs.

### Attention Mechanisms
- **Sliding window attention** (Longformer) prevents long-range incoherence by grounding each token in local context.
- **Cross-attention to structured prompts** (e.g., outlining required story beats) constrains generation to follow a logical narrative arc.
- **Controlled attention heads:** Some fine-tuning techniques (like CTRL tokens) designate specific attention heads to track stylistic constraints.

# 🌟 Exercise 4: Evaluating Trade-offs Between GANs and VAEs

## Reference: GAN vs VAE Comparison

| Dimension | GAN | VAE |
|---|---|---|
| **Output quality** | Very high (photorealistic) | Good, but often blurry |
| **Training stability** | Notoriously unstable (mode collapse, vanishing gradients) | Stable, well-defined loss |
| **Latent space structure** | Unstructured — no explicit encoder | Continuous, structured — enables interpolation |
| **Interpretability** | Low | Moderate — latent dims can be disentangled |
| **Diversity of outputs** | Risk of mode collapse | Generally more diverse |
| **Inference (encoding)** | No native encoder | Built-in encoder |
| **Mathematical basis** | Minimax game (adversarial) | ELBO (Evidence Lower Bound) |
| **Compute cost** | High | Moderate |

## 4.1 — Synthetic Medical Image Generation

**Scenario:** Generating synthetic MRI scans for training ML models while preserving privacy.

### ✅ Recommended: **GAN (specifically, a conditional GAN)**

**Justification:**
The primary requirement is **high image fidelity** — synthetic MRI scans must be anatomically realistic and indistinguishable from real scans to be useful for training downstream classifiers.

- **GANs** produce sharper, more realistic images than VAEs. A **conditional GAN** (conditioned on diagnosis label, anatomy region, patient metadata) can generate class-specific scans (e.g., tumor vs healthy).
- **VAEs** produce blurry outputs due to the pixel-level reconstruction loss, which averages over the posterior — unacceptable for medical imaging where fine detail is diagnostically significant.

**Additional consideration:** Use **FID score** to validate that generated MRI distributions match real scan distributions before using them in training. Recommended architecture: **StyleGAN2** or **Medical GAN** variants validated in the literature.

## 4.2 — AI-Assisted Creative Writing

**Scenario:** Generating text for short stories and creative writing.

### ✅ Recommended: **Neither classic GAN nor VAE — but VAE concepts apply**

**Justification:**
Text generation is dominated by **autoregressive Transformer models** (GPT family), not GANs or VAEs in their classic form. However, comparing the two architectures in principle:

- **VAEs for text** offer a structured latent space that enables semantic interpolation — you can smoothly transition between story styles, tones, or themes by manipulating the latent vector. This is valuable for a publishing tool where editors want to control narrative direction.
- **GANs for text** suffer from a fundamental problem: text is discrete, making backpropagation through the generator difficult (gradients don't flow through token sampling). Workarounds (Gumbel-softmax, SeqGAN) exist but remain unstable.

**Conclusion:** VAE-inspired latent space structure is more beneficial for creative text tools due to interpolability and control. In practice, combine with a Transformer backbone (e.g., **Optimus: BERT+GPT-2 VAE**).

## 4.3 — Anomaly Detection in Financial Transactions

**Scenario:** Detecting fraudulent transactions by learning normal purchase patterns.

### ✅ Recommended: **VAE**

**Justification:**
Anomaly detection requires the ability to **encode** a transaction and measure its reconstruction error — if a transaction is anomalous, the VAE cannot reconstruct it well from its latent representation, producing a high reconstruction loss that serves as the anomaly score.

- **VAEs have a native encoder** — perfect for this inference task.
- The structured latent space allows the model to learn a tight distribution of "normal" transactions; anomalies fall outside this distribution.
- **GANs lack a native encoder** (without modifications like BiGAN or AAE). Their discriminator only judges real vs fake — not designed for anomaly scoring on real inputs.

**Implementation:** Train VAE on legitimate transactions only. At inference, flag transactions where ELBO (reconstruction loss + KL divergence) exceeds a threshold.

## 4.4 — Generating High-Resolution Fashion Designs

**Scenario:** Creating AI-generated fashion designs and interpolating between styles.

### ✅ Recommended: **GAN for output quality; VAE for style interpolation**

**Justification:**
This task has two sub-requirements:

1. **High-resolution output quality** → **GAN wins.** StyleGAN3 can generate photorealistic clothing designs at 1024×1024+ resolution. The adversarial training forces the generator to produce sharp, fine-detailed textures — fabric grain, stitching, color gradients — that VAEs cannot match.

2. **Style interpolation** (blending between, say, a winter coat and a summer dress) → **VAE is more reliable.** Its structured latent space supports linear interpolation that produces semantically meaningful intermediate designs.

**Best of both worlds:** Use a **StyleGAN** with a learned encoder (e.g., **e4e encoder** or **pSp**) to project designs into StyleGAN's latent space, then interpolate — getting GAN quality with VAE-like control.

## 4.5 — Data Augmentation for Autonomous Vehicles

**Scenario:** Generating synthetic training data for diverse driving conditions.

### ✅ Recommended: **GAN (conditional/domain-transfer GAN)**

**Justification:**
Autonomous vehicle perception systems require photorealistic training data — rain, fog, night, construction zones, rare pedestrian configurations. Any blurriness or artifacting in synthetic data can teach the perception model incorrect features.

- **GAN advantages:** CycleGAN / pix2pix can perform domain transfer (sunny → rainy, daytime → nighttime) on real images while preserving scene geometry. This produces highly realistic augmented data.
- **VAE disadvantages:** Blurry outputs could cause object detection models to learn incorrect edge features, reducing accuracy on real driving data.

**Recommended architecture:** **CARLA + GAN pipeline** — use CARLA simulator for geometry, then apply a GAN-based style transfer to add photorealism across weather conditions.

# 🌟 Exercise 5: Advanced Latent Space Exploration in VAEs

This exercise includes both theoretical analysis and live code experiments on the MNIST dataset.

## Setup — Install & Import Dependencies

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
from torchvision import datasets, transforms

import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

## VAE Architecture

```python
class VAE(nn.Module):
    def __init__(self, input_dim=784, hidden_dim=512, latent_dim=2):
        super(VAE, self).__init__()
        self.latent_dim = latent_dim

        # Encoder
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, 256),
            nn.ReLU()
        )
        self.fc_mu = nn.Linear(256, latent_dim)
        self.fc_logvar = nn.Linear(256, latent_dim)

        # Decoder
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, 256),
            nn.ReLU(),
            nn.Linear(256, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, input_dim),
            nn.Sigmoid()
        )

    def encode(self, x):
        h = self.encoder(x)
        return self.fc_mu(h), self.fc_logvar(h)

    def reparameterize(self, mu, logvar):
        """Reparameterization trick: z = mu + eps * std"""
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std

    def decode(self, z):
        return self.decoder(z)

    def forward(self, x):
        mu, logvar = self.encode(x)
        z = self.reparameterize(mu, logvar)
        x_recon = self.decode(z)
        return x_recon, mu, logvar


def vae_loss(x_recon, x, mu, logvar, beta=1.0):
    """
    ELBO Loss = Reconstruction Loss + beta * KL Divergence
    beta=1: standard VAE
    beta>1: beta-VAE (encourages disentanglement)
    """
    recon_loss = F.binary_cross_entropy(x_recon, x, reduction='sum')
    kl_loss = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp())
    return recon_loss + beta * kl_loss, recon_loss.item(), kl_loss.item()


print("VAE architecture defined.")
```

## Training the VAE on MNIST

```python
# Hyperparameters
LATENT_DIM = 2     # 2D for visualization; increase to 10+ for better quality
BATCH_SIZE = 256
EPOCHS = 15
LR = 1e-3
BETA = 1.0         # Try beta=4 for beta-VAE disentanglement (Ex 5.3)

# Data
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Lambda(lambda x: x.view(-1))  # Flatten to 784
])

train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset  = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader  = DataLoader(test_dataset,  batch_size=BATCH_SIZE, shuffle=False)

# Model
model = VAE(latent_dim=LATENT_DIM).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=LR)

# Training loop
train_losses = []

for epoch in range(1, EPOCHS + 1):
    model.train()
    total_loss = 0
    for x_batch, _ in train_loader:
        x_batch = x_batch.to(device)
        optimizer.zero_grad()
        x_recon, mu, logvar = model(x_batch)
        loss, _, _ = vae_loss(x_recon, x_batch, mu, logvar, beta=BETA)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    avg_loss = total_loss / len(train_dataset)
    train_losses.append(avg_loss)
    print(f"Epoch {epoch:02d}/{EPOCHS} — Loss: {avg_loss:.4f}")

print("\n✅ Training complete.")
```

## 5.1 — Visualizing Latent Space Distributions

### Theory
A well-trained VAE should encode different digit classes into **distinct, well-separated regions** of the latent space. We evaluate this by:
1. **Visual inspection** — encode the test set and color points by digit class.
2. **Silhouette score** — measures how well-separated clusters are (range -1 to 1; higher is better).
3. **t-SNE / UMAP** — non-linear dimensionality reduction to visualize high-dimensional latent spaces.

With a 2D latent space, we can plot directly without t-SNE.

## 5.2 — Interpolating Between Two Samples

### Theory
**Why interpolation works in VAEs but not (natively) in GANs:**

- **VAE:** The encoder maps any input to a point in a *continuous, structured* latent space. By design (KL divergence regularization), the latent space is smooth — nearby points decode to semantically similar images. A linear interpolation `z = α·z₁ + (1-α)·z₂` traverses a meaningful path.

- **GAN:** GANs have no encoder. To interpolate, you'd need to find the latent codes `z₁` and `z₂` that correspond to specific images via optimization ("GAN inversion") — an expensive, unstable process. Even with inversion, GAN latent spaces are not regularized to be smooth, so interpolated points may decode to unrealistic images.

**Process:**
1. Encode image `A` (digit '3') → `(μ_A, σ_A)` → sample `z_A`
2. Encode image `B` (digit '8') → `(μ_B, σ_B)` → sample `z_B`
3. Generate interpolated points: `z_t = (1-t)·z_A + t·z_B` for `t ∈ [0, 1]`
4. Decode each `z_t` to produce intermediate images

```python
def get_sample_from_class(dataset, target_class):
    """Return a single sample from the specified digit class."""
    for img, label in dataset:
        if label == target_class:
            return img.to(device)
    return None


model.eval()
with torch.no_grad():
    # Get one example of digit '3' and digit '8'
    img_3 = get_sample_from_class(test_dataset, target_class=3).unsqueeze(0)
    img_8 = get_sample_from_class(test_dataset, target_class=8).unsqueeze(0)

    mu_3, _ = model.encode(img_3)
    mu_8, _ = model.encode(img_8)

    # Interpolate
    n_steps = 10
    alphas = np.linspace(0, 1, n_steps)
    interpolated_imgs = []

    for alpha in alphas:
        z_interp = (1 - alpha) * mu_3 + alpha * mu_8
        img_interp = model.decode(z_interp).cpu().view(28, 28).numpy()
        interpolated_imgs.append(img_interp)

# Plot
fig, axes = plt.subplots(1, n_steps, figsize=(20, 2.5))
for i, (ax, img) in enumerate(zip(axes, interpolated_imgs)):
    ax.imshow(img, cmap='gray')
    ax.axis('off')
    if i == 0:
        ax.set_title("'3'", fontsize=10)
    elif i == n_steps - 1:
        ax.set_title("'8'", fontsize=10)
    else:
        ax.set_title(f"α={alphas[i]:.1f}", fontsize=8)

plt.suptitle("Smooth Interpolation in VAE Latent Space: '3' → '8'", fontsize=13, y=1.05)
plt.tight_layout()
plt.show()
```

## 5.3 — Controlling the Degree of Variability (KL Divergence Weight)

### Theory

The VAE loss is:
$$\mathcal{L} = \underbrace{\mathbb{E}[\log p(x|z)]}_{\text{Reconstruction}} - \underbrace{\beta \cdot D_{KL}(q(z|x) \| p(z))}_{\text{KL Regularization}}$$

| β value | Effect on KL | Effect on latent space | Effect on output |
|---|---|---|---|
| β = 0 | KL ignored | Posterior collapses — encodes every nuance | Very sharp reconstruction, no generation ability |
| β = 1 | Standard VAE | Balanced structure | Good quality + reasonable diversity |
| β > 1 (β-VAE) | Strong regularization | Compressed, well-organized | More disentangled but potentially blurrier |
| β >> 1 | Overregularized | Posterior ≈ prior (N(0,I)) | Generic, uninformative outputs |

**Real-world application where variability control matters:**
In drug molecule generation, you want controlled variability:
- Too low β → generates molecules that memorize training data (no novelty)
- Too high β → generates random molecular graphs that are chemically invalid
- Optimal β → generates novel but chemically plausible molecules near known compounds

This is exactly the use case of **JTVAE (Junction Tree VAE)** for molecular generation in pharmaceutical research.

```python
def train_vae_with_beta(beta, epochs=10):
    """Train a fresh VAE with specified beta and return the model + final loss."""
    m = VAE(latent_dim=LATENT_DIM).to(device)
    opt = torch.optim.Adam(m.parameters(), lr=LR)
    for epoch in range(epochs):
        m.train()
        for x_batch, _ in train_loader:
            x_batch = x_batch.to(device)
            opt.zero_grad()
            x_recon, mu, logvar = m(x_batch)
            loss, _, _ = vae_loss(x_recon, x_batch, mu, logvar, beta=beta)
            loss.backward()
            opt.step()
    return m


def sample_and_show(model_dict, n_samples=10):
    """Sample from prior N(0,I) and decode for each beta model."""
    fig, axes = plt.subplots(len(model_dict), n_samples, figsize=(n_samples * 1.5, len(model_dict) * 1.5))

    for row_idx, (beta_label, m) in enumerate(model_dict.items()):
        m.eval()
        with torch.no_grad():
            z = torch.randn(n_samples, LATENT_DIM).to(device)
            samples = m.decode(z).cpu().view(n_samples, 28, 28).numpy()

        for col_idx in range(n_samples):
            axes[row_idx, col_idx].imshow(samples[col_idx], cmap='gray')
            axes[row_idx, col_idx].axis('off')
        axes[row_idx, 0].set_ylabel(beta_label, fontsize=11, rotation=0, labelpad=60)

    plt.suptitle("Effect of β on VAE Sample Quality and Diversity", fontsize=13)
    plt.tight_layout()
    plt.show()


print("Training VAEs with different β values (this will take a few minutes)...")
betas = {"β = 0.1 (low KL)": 0.1, "β = 1.0 (standard)": 1.0, "β = 8.0 (β-VAE)": 8.0}
trained_models = {label: train_vae_with_beta(b, epochs=10) for label, b in betas.items()}
sample_and_show(trained_models)
```

## 5.4 — Disentangling Latent Representations

### Theory
A **disentangled latent space** means each dimension `z_i` independently controls one interpretable factor of variation (e.g., `z_0` = smile intensity, `z_1` = hair color, `z_2` = age).

**How to modify only one attribute (e.g., expression) while keeping others fixed:**
1. Encode the target face → get `z = [z_0, z_1, ..., z_n]`
2. Identify which dimension corresponds to the desired attribute:
   - **Supervised:** Train a linear classifier on labeled data (smile / no smile) using latent codes; the most discriminative dimension is the smile dimension.
   - **Unsupervised (β-VAE):** Use a high β value to pressure the model into disentangling; then traverse each dimension independently and observe which attribute changes.
3. Modify only `z_k` (the identified attribute dimension) while holding all other dimensions constant.
4. Decode the modified `z` → the output changes only in the targeted attribute.

**Example:** For a face generator trained with β-VAE on CelebA:
- `z_3 ± Δ` → changes smile
- `z_7 ± Δ` → changes hair darkness
- `z_1 ± Δ` → changes age

This is only approximately possible — in practice, some entanglement always remains.

```python
# Demonstrate latent traversal on MNIST with 2D latent space
# Traverse each latent dimension independently to see what it controls

model.eval()

fig, axes = plt.subplots(2, 10, figsize=(20, 4))
fig.suptitle('Latent Dimension Traversal — Controlling One Dimension at a Time', fontsize=12)

traversal_range = np.linspace(-3, 3, 10)

for dim_idx in range(LATENT_DIM):
    for col_idx, val in enumerate(traversal_range):
        z = torch.zeros(1, LATENT_DIM).to(device)
        z[0, dim_idx] = val  # Only vary this dimension; others stay at 0 (prior mean)

        with torch.no_grad():
            img = model.decode(z).cpu().view(28, 28).numpy()

        axes[dim_idx, col_idx].imshow(img, cmap='gray')
        axes[dim_idx, col_idx].axis('off')
        if col_idx == 0:
            axes[dim_idx, col_idx].set_ylabel(f'z_{dim_idx}', fontsize=12)
        if dim_idx == 0:
            axes[dim_idx, col_idx].set_title(f'{val:.1f}', fontsize=8)

plt.tight_layout()
plt.show()
print("Each row shows what changes when we vary one latent dimension while the other is held at 0.")
```

## 5.5 — Comparing PCA and VAEs

### Theoretical Comparison

| Dimension | PCA | VAE |
|---|---|---|
| **Linearity** | Strictly linear — finds orthogonal axes of maximum variance in original space | Non-linear — encoder is a deep neural network that captures complex manifolds |
| **Interpretability** | High — each principal component is a linear combination of original features with known explained variance ratio | Lower — latent dimensions are abstract; require traversal experiments to interpret |
| **Reconstruction quality** | Good for data close to linear subspaces; fails on complex non-linear data (images, speech) | Excellent — can reconstruct complex non-linear patterns |
| **Generative modeling** | Cannot generate new samples — no probabilistic prior; PCA inverse transform only interpolates | Fully generative — sample `z ~ N(0,I)`, decode to new samples |
| **Scalability** | O(d²) — struggles with very high-dimensional data | Scales with network size — handles high-res images |
| **Training** | Closed-form SVD — no gradient-based training needed | Requires SGD training, hyperparameter tuning |

```python
# Side-by-side comparison: PCA vs VAE latent space on MNIST

# Load raw test images
test_images, test_labels_raw = [], []
for img, label in test_dataset:
    test_images.append(img.numpy())
    test_labels_raw.append(label)
    if len(test_images) >= 3000:
        break

test_images_np = np.stack(test_images)         # shape: (3000, 784)
test_labels_np = np.array(test_labels_raw)

# PCA to 2D
pca = PCA(n_components=2)
pca_result = pca.fit_transform(test_images_np)

# VAE encoding to 2D (already computed as all_mu above — use first 3000)
vae_result = all_mu[:3000]

# Plot
fig, axes = plt.subplots(1, 2, figsize=(16, 7))

for ax, data, title in zip(
    axes,
    [pca_result, vae_result],
    ['PCA — 2D Projection of MNIST', 'VAE — 2D Latent Space of MNIST']
):
    scatter = ax.scatter(data[:, 0], data[:, 1],
                         c=test_labels_np, cmap='tab10', alpha=0.5, s=8)
    ax.set_title(title, fontsize=12)
    ax.set_xlabel('Dim 1')
    ax.set_ylabel('Dim 2')
    plt.colorbar(scatter, ax=ax, label='Digit')

plt.tight_layout()
plt.show()

print(f"PCA explained variance: {pca.explained_variance_ratio_.sum()*100:.1f}%")
print("Observation: PCA projects linearly → classes overlap more.")
print("VAE learns non-linear manifolds → classes are better separated in latent space.")
```

```python
# Bonus: Reconstruction quality comparison — PCA vs VAE

n_show = 8
sample_imgs = test_images_np[:n_show]  # (n_show, 784)

# PCA reconstruction
pca_full = PCA(n_components=2).fit(test_images_np)
pca_encoded = pca_full.transform(sample_imgs)
pca_recon = pca_full.inverse_transform(pca_encoded)

# VAE reconstruction
model.eval()
with torch.no_grad():
    x_tensor = torch.tensor(sample_imgs, dtype=torch.float32).to(device)
    vae_recon, _, _ = model(x_tensor)
    vae_recon = vae_recon.cpu().numpy()

fig, axes = plt.subplots(3, n_show, figsize=(n_show * 2, 6))
row_labels = ['Original', 'PCA Recon', 'VAE Recon']
row_data   = [sample_imgs, pca_recon, vae_recon]

for row_idx, (label, data) in enumerate(zip(row_labels, row_data)):
    for col_idx in range(n_show):
        axes[row_idx, col_idx].imshow(data[col_idx].reshape(28, 28), cmap='gray')
        axes[row_idx, col_idx].axis('off')
    axes[row_idx, 0].set_ylabel(label, fontsize=11, rotation=0, labelpad=70)

plt.suptitle('Reconstruction Quality: PCA vs VAE (2D latent)', fontsize=13)
plt.tight_layout()
plt.show()
```
