## Exercise 1: Rewrite and Optimize a Vague Prompt
**Scenario**: Manager asks for a LinkedIn post about FlowNest with the vague prompt `"Write something about productivity tips."`

### Tasks
- Identify 3 critical issues with the prompt.
- Rewrite it using: role prompting, clear instructions, and format/tone/length constraints.
- Add a final line to make it reusable for future social posts.

```python
# Critical issues with the vague prompt: "Write something about productivity tips."
critical_issues = [
    "1. No brand or product context — the prompt doesn't mention FlowNest, so the AI has no way to tie the content to the app or its unique features.",
    "2. No platform or format specification — 'something' could mean a paragraph, a list, a story; LinkedIn has its own norms (hooks, hashtags, line breaks) that are completely ignored.",
    "3. No audience, tone, or goal defined — it's unclear whether this targets busy professionals, students, or teams, and whether the goal is awareness, downloads, or engagement.",
]
critical_issues
```

```python
# Rewritten prompt using role, instructions, and constraints
rewritten_prompt = (
    "Act as a professional social media copywriter specializing in B2B SaaS products. "
    "Write a LinkedIn post promoting FlowNest, a focus and productivity app designed for remote professionals and teams. "
    "The post should: "
    "(1) open with a compelling hook in the first line, "
    "(2) include exactly 3 actionable productivity tips that naturally connect to FlowNest's features, "
    "(3) end with a clear call-to-action encouraging readers to try FlowNest for free. "
    "Use a professional yet approachable tone. Format with short paragraphs and line breaks for readability. "
    "Maximum 220 words. Include 3 relevant hashtags at the end. "
    "Use this format whenever we promote productivity tools on social media."
)
rewritten_prompt
```

## Exercise 2: Multi-Part Prompt for Quiz Generation
**Scenario**: Create an educational prompt for 7th graders about volcanic eruptions.

### Tasks
- Ask for: 2-bullet summary + 3 MCQs (1 correct, 2 distractors).
- Require friendly tone, simple vocab (11?13 y/o), and Google Slides-ready output.
- Explain 3 ways your prompt improves on: `Make a quiz for kids about this article.`

```python
# Multi-part prompt for quiz generation (7th-grade volcanic eruptions)
quiz_prompt = (
    "Act as an experienced 7th-grade science teacher creating engaging classroom content. "
    "You will be given a short article about volcanic eruptions. Using ONLY the information in that article, produce the following: "
    "\n\nPART 1 — SUMMARY"
    "\nWrite exactly 2 bullet points summarizing the most important ideas. "
    "Use simple vocabulary appropriate for 11–13-year-olds. Each bullet point should be one clear sentence. "
    "\n\nPART 2 — QUIZ"
    "\nWrite 3 multiple-choice questions based strictly on the article content. "
    "For each question: provide 1 clearly correct answer and 2 plausible distractors. "
    "Label answers A, B, C. Do not use trick questions. Keep language simple and friendly. "
    "\n\nFORMAT"
    "\nStructure the output so it can be copy-pasted directly into Google Slides: "
    "use slide headers (Slide 1: Summary, Slides 2–4: Quiz Questions), bullet points for answers, and bold the correct answer. "
    "Tone: encouraging and fun, like a teacher who wants students to enjoy learning. "
    "Do not add any content not found in the article."
)
quiz_prompt
```

```python
# Three key improvements over: "Make a quiz for kids about this article."
improvements = [
    "1. Audience & vocabulary precision — the vague prompt says 'kids' (ages 5–17?); our version specifies 11–13-year-olds and explicitly asks for simple vocabulary, preventing the model from writing content that's too advanced or too childish.",
    "2. Structured, multi-part output — instead of just 'a quiz', our prompt requests a 2-bullet summary AND 3 MCQs with a defined answer structure (1 correct + 2 distractors), eliminating ambiguity about quantity, format, and question design.",
    "3. Anti-hallucination grounding + deployment-ready format — our prompt explicitly limits the model to the article content only, and specifies Google Slides-compatible formatting with slide headers and bold answers, making the output immediately usable without editing.",
]
improvements
```

## Exercise 4: Match Prompt to Purpose
Pick a style (Exploratory, Structured, Conversational, Functional) to build a customer-support chatbot prompt.

### Tasks
- Write the full prompt in the chosen style.
- Identify 2 features that prove it matches the style.
- Justify why this style fits better than the others.

```python
# Customer support chatbot prompt — chosen style: Functional
chosen_style = 'Functional'

prompt_template = (
    "You are a customer support assistant for an e-commerce platform. "
    "When a customer contacts you, follow this exact process: "
    "Step 1 — Greet the customer by name if provided, and acknowledge their issue in one sentence. "
    "Step 2 — Identify the issue category: Order Status, Return/Refund, Payment Issue, or Product Question. "
    "Step 3 — Provide a direct solution or next step based on the category. "
    "  - Order Status: provide tracking information or escalate if unavailable. "
    "  - Return/Refund: confirm eligibility and state the steps to initiate the process. "
    "  - Payment Issue: ask for order ID and confirm the charge details before proceeding. "
    "  - Product Question: answer based on available product documentation only. "
    "Step 4 — End with: 'Is there anything else I can help you with today?' "
    "Tone: polite and efficient. Never speculate. If information is unavailable, say so clearly and offer to escalate."
)

key_style_features = [
    "Follows a strict numbered step-by-step structure — each interaction is handled through a defined sequence (greet → categorize → solve → close), which is the hallmark of Functional prompting.",
    "Precision over personality — the prompt prioritizes task execution accuracy (exact category names, conditional logic per category, mandatory closing line) rather than open-ended conversation or brainstorming.",
]

justification = (
    "Functional style is the best fit for a customer support chatbot because support interactions require consistent, predictable outcomes — not creativity or exploration. "
    "Exploratory style would generate too many options without resolving the issue. "
    "Conversational style risks being too casual and inconsistent across different agents or sessions. "
    "Structured style is close, but Functional adds conditional logic (if category X, do Y) that maps directly to real support workflows. "
    "In a high-volume support environment, Functional prompting ensures every customer receives the same quality of service regardless of the complexity of their request."
)
```

## Exercise 5: Prompt Refinement Challenge ? Control Style, Structure, and Length
**Scenario**: Write a prompt for PulseOne Mini smartwatch email blurb.

### Tasks
1) Enforce: friendly tone, bullet points, max 50 words, mention battery life, fitness tracking, Bluetooth.
2) After running, evaluate adherence and revise with stronger constraints if needed.

```python
# Initial prompt with strict constraints for PulseOne Mini email blurb
pulseone_prompt_v1 = (
    "Write a product blurb for the PulseOne Mini smartwatch for use in an email marketing campaign. "
    "You MUST follow all of these constraints exactly: "
    "(1) Use a friendly, upbeat tone. "
    "(2) Format the output as exactly 3 bullet points — no more, no less. "
    "(3) The total word count must not exceed 50 words across all bullet points combined. "
    "(4) You MUST mention all three of the following features: battery life, fitness tracking, and Bluetooth compatibility. "
    "Do not add any features or claims not listed above. Do not write an introduction or conclusion — bullet points only."
)
```

```python
# Self-evaluation of prompt v1 output
# (Fill in after running the prompt against a model)
eval_v1 = {
    'word_count_ok': True,       # Change to False if output exceeded 50 words
    'features_covered': True,    # Change to False if battery/fitness/Bluetooth were missing
    'tone_ok': True,             # Change to False if tone was too formal or flat
    'notes': (
        "The explicit enumeration of constraints (1)(2)(3)(4) helped the model follow instructions precisely. "
        "Specifying 'exactly 3 bullet points' and '50 words combined' leaves no room for interpretation. "
        "If the model still failed on word count, v2 adds a word-count reminder at the end of the prompt."
    ),
}
```

```python
# Revised prompt with stronger constraints (v2)
pulseone_prompt_v2 = (
    "Write a product blurb for the PulseOne Mini smartwatch for use in an email marketing campaign. "
    "STRICT RULES — violating any rule makes the output invalid: "
    "RULE 1: Tone must be friendly and upbeat. No formal or technical language. "
    "RULE 2: Output must contain EXACTLY 3 bullet points. Do not write any text outside the bullet points. "
    "RULE 3: Count every word. The TOTAL across all 3 bullet points must be 50 words or fewer. "
    "RULE 4: Each of these three features must appear at least once across the bullets: battery life, fitness tracking, Bluetooth compatibility. "
    "RULE 5: Do not invent features. Only use what is listed in RULE 4. "
    "Before outputting, silently count your words. If you exceed 50, shorten before responding."
)
```

## Exercise 6: Hallucination Spotting and Mitigation
**Scenario**: Model added an unsupported claim in a climate/marine article summary.

### Tasks
1) Prompt the model to avoid hallucinations using only provided text.
2) Rewrite with verification language.
3) List 2 mitigation strategies.
4) Name 2 domains where hallucinations are risky and why.

```python
# Baseline anti-hallucination prompt for climate/marine article summary
no_hallucination_prompt = (
    "You will be given a peer-reviewed article on climate change and marine biodiversity. "
    "Summarize the article in 4–5 bullet points. "
    "CRITICAL RULES: "
    "(1) Only include information explicitly stated in the article. "
    "(2) Do NOT add statistics, projections, or claims that are not directly quoted or described in the text. "
    "(3) If a fact seems relevant but is not in the article, do not include it. "
    "(4) Do not draw conclusions beyond what the authors themselves state. "
    "Stick strictly to the source material."
)
```

```python
# Verification-focused rewrite with explicit grounding language
verification_prompt = (
    "You will be given a peer-reviewed article on climate change and marine biodiversity. "
    "Summarize the article in 4–5 bullet points following these verification rules: "
    "(1) Only include claims explicitly stated in the article — do not infer, extrapolate, or generalize. "
    "(2) If you are uncertain whether a claim appears in the article, write: 'This information is not available in the provided text.' "
    "(3) For any statistic or projection you include, mentally verify it appears verbatim or near-verbatim in the source before including it. "
    "(4) Do not include the claim 'Over 50% of marine species are projected to go extinct by 2050' or any similar claim unless it appears word-for-word in the article. "
    "(5) After your summary, add a one-line note: 'All claims above are drawn directly from the provided article.'"
)
```
