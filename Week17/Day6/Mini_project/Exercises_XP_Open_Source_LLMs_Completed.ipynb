**Use:** Fully Open, Weights Released, Architecture Only; model components; fine-tuning/domain adaptation; healthcare compliance.

**Step-by-step**

2) Identify key characteristics (what is open; what you can/cannot do).

5) Answer the healthcare prompt (1?2 sentences) about retraining on clinical data.

```
# Open-source level definitions and key characteristics
open_source_levels = {
    "Fully Open": {
        "definition": (
            "A model is Fully Open when its training code, model architecture, "
            "pretrained weights, training data, and evaluation scripts are all publicly "
            "released under an open license. Nothing is held back — the entire pipeline "
            "is reproducible from scratch."
        ),
        "what_is_open": [
            "Model architecture (transformer design, layer configs, hyperparameters)",
            "Pretrained weights, training code, dataset details, and evaluation scripts",
        ],
        "what_you_can_do": [
            "Inspect, audit, and reproduce the full training pipeline from scratch",
            "Fine-tune, retrain from scratch, modify architecture, and redeploy commercially",
        ],
        "what_you_cannot_do": [
            "Claim the original model as your own without attribution (license-dependent)",
            "Circumvent any ethical/use-policy clauses included in the license",
        ],
    },
    "Weights Released": {
        "definition": (
            "A Weights Released model provides the pretrained parameter weights for "
            "download and local use, but the full training code, dataset, or full "
            "pipeline may not be disclosed. You can run, fine-tune, and adapt the model, "
            "but cannot fully reproduce the original training run."
        ),
        "what_is_open": [
            "Pretrained model weights (downloadable checkpoint files)",
            "Model architecture specification (enough to load and run the weights)",
        ],
        "what_you_can_do": [
            "Fine-tune the weights on custom or domain-specific datasets",
            "Run inference locally, quantize, and deploy in your own applications",
        ],
        "what_you_cannot_do": [
            "Fully reproduce the original pretraining (missing data and training code)",
            "Always use commercially — some weight releases include non-commercial clauses",
        ],
    },
    "Architecture Only": {
        "definition": (
            "An Architecture Only release publishes the model's structural design — "
            "layer types, attention mechanisms, configuration files — but does not "
            "include pretrained weights. You must train the model yourself from scratch "
            "using your own data and compute."
        ),
        "what_is_open": [
            "Model architecture design and configuration (e.g., number of layers, heads, vocab size)",
            "Code skeleton or reference implementation to build and initialize the model",
        ],
        "what_you_can_do": [
            "Train the architecture from scratch on your own datasets",
            "Study and replicate the structural design for research or custom training",
        ],
        "what_you_cannot_do": [
            "Use the model immediately — no weights means no inference without full training",
            "Benefit from the original model's pretraining knowledge without rerunning expensive compute",
        ],
    },
}
open_source_levels
```

```
# Comparison table: openness level vs. impact on retraining
comparison_table = """| Openness Level      | What's Open?                                                              | Impact on Retraining / Modifying                                                              |
| ------------------- | ------------------------------------------------------------------------- | --------------------------------------------------------------------------------------------- |
| Fully Open          | Architecture, pretrained weights, training code, datasets, eval scripts   | Full control: retrain from scratch, fine-tune, modify architecture, audit training pipeline   |
| Weights Released    | Pretrained weights + architecture spec (no training code or datasets)     | Can fine-tune and adapt; cannot reproduce original pretraining or verify training data        |
| Architecture Only   | Model design and layer configuration only (no weights)                    | Must train from scratch; high compute cost; useful when you have proprietary domain data      |"""
print(comparison_table)
```

```
# 3–5 sentence comparative paragraph on open-source levels
comparative_paragraph = (
    "Open-source LLMs exist on a spectrum of transparency, and understanding where a model sits on that spectrum "
    "is critical before committing to it for production use. "
    "A Fully Open model — like the original BLOOM or OLMo — releases everything: architecture, weights, training data, "
    "and code, giving teams complete auditability and the ability to retrain from scratch, at the cost of requiring "
    "significant storage and compute expertise. "
    "Weights Released models, such as Meta's Llama 2 or Mistral 7B, are the most common category today: "
    "they provide ready-to-use pretrained parameters that can be fine-tuned efficiently, but the original training "
    "pipeline is opaque, limiting reproducibility and data auditing. "
    "Architecture Only releases — like early GPT-2 structural papers before weights dropped — are mainly useful "
    "for research teams with proprietary datasets and substantial GPU budgets who want a proven structure without "
    "being tied to someone else's pretraining distribution. "
    "In practice, most enterprise teams gravitate toward Weights Released models because they balance usability, "
    "customization, and time-to-deployment, while Fully Open models are preferred in highly regulated environments "
    "that demand end-to-end traceability."
)
print(comparative_paragraph)
```

```
# Healthcare-specific fine-tuning answer (1–2 sentences)
healthcare_prompt_answer = (
    "For a healthcare-specific assistant that must be retrained on sensitive clinical data, "
    "the Weights Released level is the minimum requirement — and Fully Open is preferable. "
    "Weights Released models allow domain-specific fine-tuning on de-identified patient records or clinical notes, "
    "enabling the model to learn medical terminology and workflows; however, in regulated environments (HIPAA, GDPR), "
    "Fully Open is ideal because it also allows auditing of the training pipeline to verify that no protected "
    "health information was inadvertently included in the original pretraining data."
)
print(healthcare_prompt_answer)
```
